```python
import os
import pandas as pd
import numpy as np

def standardize_sequence(data, max_frames=200):
    """Pads or truncates data to a fixed length."""
    if data.shape[0] > max_frames:
        return data[:max_frames, :]
    else:
        padding = np.zeros((max_frames - data.shape[0], data.shape[1]))
        return np.vstack((data, padding))

def universal_loader(root_dir):
    X = []
    y = []

    for root, dirs, files in os.walk(root_dir):
        for file in files:
            file_path = os.path.join(root, file)

            label = 1 if 'correct' in root.lower() else 0

            try:
                if file.endswith('.csv'):
                    df = pd.read_csv(file_path)
                elif file.endswith('.json'):
                    df = pd.read_json(file_path)
                else:
                    continue

                # Extract numeric coordinate data (X, Y, Z)
                # Ensure you select only coordinates/angles, not timestamps or headers
                data = df.select_dtypes(include=[np.number]).values

                # Standardize to 200 frames
                seq = standardize_sequence(data)

                X.append(seq)
                y.append(label)

            except Exception as e:
                print(f"Skipping {file}: {e}")

    return np.array(X), np.array(y)

X_data, y_data = universal_loader('/content/drive/MyDrive/datasets')

print(f"Loading complete!")
print(f"Input Shape (Total Reps, Frames, Features): {X_data.shape}")
```

```python
import os
import numpy as np
import pandas as pd

BASE_DIR = '/content/drive/MyDrive/datasets'

JOINT_MAP = {
    'l_hip':23,'l_knee':28,'l_ankle':30,
    'r_hip':24,'r_knee':34,'r_ankle':36
}
JOINT_ORDER = ['l_hip','l_knee','l_ankle','r_hip','r_knee','r_ankle']
MOVEMENT_MAP = {
    'm01':'deep_squat','m02':'hurdle_step','m03':'inline_lunge',
    'm04':'side_lunge','m05':'sit_to_stand','m06':'straight_leg_raise'
}
LOWER_LIMB = {'m01','m02','m03','m04','m05','m06'}

rows = []
for label_name, label_val, suffix in [
    ('Segmented Movements', 1, ''),
    ('Incorrect Segmented Movements', 0, '_inc'),
]:
    pos_dir = os.path.join(BASE_DIR, label_name, 'Vicon', 'Positions')
    for fname in sorted(os.listdir(pos_dir)):
        if not fname.endswith('.txt'): continue
        parts = fname.replace('_positions','').replace(suffix,'') \
                     .replace('.txt','').split('_')
        if len(parts) < 3: continue
        mov, subj, ex = parts[0], parts[1], parts[2]
        if mov not in LOWER_LIMB: continue

        df  = pd.read_csv(os.path.join(pos_dir, fname), header=None)
        arr = df.values.astype(np.float32)
        T   = len(arr)

        for t in range(T):
            row = {
                'subject':   subj,
                'movement':  mov,
                'exercise':  MOVEMENT_MAP[mov],
                'exercise_id': ex,
                'frame':     t,
                'label':     label_val,
            }
            for ji, jname in enumerate(JOINT_ORDER):
                c = JOINT_MAP[jname] * 3
                row[f'{jname}_x'] = arr[t, c]
                row[f'{jname}_y'] = arr[t, c+1]
                row[f'{jname}_z'] = arr[t, c+2]
            rows.append(row)

master_df = pd.DataFrame(rows)
print("Shape:", master_df.shape)
print("Columns:", master_df.columns.tolist())
print(master_df.head(3))

# save
out_path = '/content/drive/MyDrive/datasets/UIPRMD_timeseries.csv'
master_df.to_csv(out_path, index=False)
print(f"\nSaved to {out_path}")
print(f"Subjects: {master_df['subject'].nunique()}")
print(f"Exercises: {master_df['exercise'].unique()}")
print(f"Correct: {(master_df['label']==1).sum()} frames")
print(f"Incorrect: {(master_df['label']==0).sum()} frames")
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d
from typing import Dict, List, Optional, Tuple

# â”€â”€ config â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

TARGET_LEN  = 150
N_JOINTS    = 6
IN_CHANNELS = 3
MIN_FRAMES  = 20
FPS         = 30

UIPRMD_CSV = '/content/drive/MyDrive/datasets/UIPRMD_timeseries.csv'
KIMORE_CSV = '/content/drive/MyDrive/datasets/squats_tabular_timeseries_binary.csv'
KERAAL_CSV = '/content/drive/MyDrive/datasets/keraal_final_binary.csv'

JOINT_ORDER = ['l_hip','l_knee','l_ankle','r_hip','r_knee','r_ankle']

# â”€â”€ hardware modes â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

HARDWARE_MODES = {
    0: {'name': 'free',              'scale': 0.0,  'bias': 0.0,  'min': 0.0, 'max': 0.0},
    1: {'name': 'constant_torque',   'scale': 0.0,  'bias': 0.5,  'min': 0.4, 'max': 0.6},
    2: {'name': 'resistive_torque',  'scale': 0.5,  'bias': 0.0,  'min': 0.0, 'max': 0.8},
    3: {'name': 'increasing_torque', 'scale': 1.0,  'bias': 0.0,  'min': 0.0, 'max': 1.0},
    4: {'name': 'dampening',         'scale': -0.5, 'bias': 0.5,  'min': 0.0, 'max': 0.5},
}
N_MODES = len(HARDWARE_MODES)

def compute_torque(quality: float, mode: int,
                   velocity_proxy: float = 0.0) -> float:
    m = HARDWARE_MODES.get(mode, HARDWARE_MODES[0])
    if mode == 4:
        torque = m['bias'] - abs(velocity_proxy) * 0.5
    else:
        torque = m['scale'] * quality + m['bias']
    return float(np.clip(torque, m['min'], m['max']))

# â”€â”€ adjacency â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def build_adjacency() -> np.ndarray:
    A = np.zeros((N_JOINTS, N_JOINTS), dtype=np.float32)
    for i, j in [(0,1),(1,2),(3,4),(4,5),(0,3),(1,4),(2,5)]:
        A[i,j] = A[j,i] = 1.0
    np.fill_diagonal(A, 1.0)
    d = np.diag(1.0 / np.sqrt(A.sum(axis=1) + 1e-6))
    return (d @ A @ d).astype(np.float32)

ADJ = build_adjacency()

EPS = 1e-6

def safe_np(x, fill=0.0, clip=None, dtype=np.float32):
    x = np.asarray(x, dtype=dtype)
    x = np.nan_to_num(x, nan=fill, posinf=fill, neginf=fill)
    if clip is not None:
        x = np.clip(x, -clip, clip)
    return x.astype(dtype)

def valid_np(x) -> bool:
    x = np.asarray(x)
    return x.size > 0 and np.isfinite(x).all()

# â”€â”€ signal helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def _smooth(arr, win=7, poly=2):
    arr = safe_np(arr)
    if len(arr) < win + 2:
        return arr
    w = win if win % 2 == 1 else win + 1
    w = max(w, poly+2 if (poly+2)%2==1 else poly+3)
    try:
        return safe_np(savgol_filter(arr, w, poly))
    except Exception:
        return arr

def _resample(arr, n):
    arr = safe_np(arr)
    if len(arr) == n:
        return arr
    if len(arr) < 2:
        return np.zeros(n, dtype=np.float32)
    return safe_np(np.interp(np.linspace(0,1,n), np.linspace(0,1,len(arr)), arr))

def _fill_missing(kps):
    kps = safe_np(kps)
    T, J, C = kps.shape
    for j in range(J):
        for c in range(C):
            col = kps[:,j,c]
            z   = (~np.isfinite(col)) | (col == 0.0)
            if z.all():
                opp = j+3 if j < 3 else j-3
                kps[:,j,c] = kps[:,opp,c]
            elif z.any() and (~z).sum() >= 2:
                f = interp1d(np.where(~z)[0], col[~z],
                             bounds_error=False, fill_value='extrapolate')
                kps[:,j,c] = f(np.arange(T))
    return safe_np(kps)

def _normalise(kps):
    kps = safe_np(kps)
    hip   = ((kps[:,0,:] + kps[:,3,:]) / 2.0).mean(axis=0)
    kps   = kps - hip
    l     = np.mean(np.linalg.norm(kps[:,0,:] - kps[:,2,:], axis=1))
    r     = np.mean(np.linalg.norm(kps[:,3,:] - kps[:,5,:], axis=1))
    scale = (l + r) / 2.0
    if (not np.isfinite(scale)) or scale < 1e-4:
        scale = 1.0
    kps   = kps / scale
    kps   = np.clip(kps, -5.0, 5.0)
    return safe_np(kps, clip=5.0)

def process_keypoints(raw: np.ndarray) -> np.ndarray:
    raw = safe_np(raw)
    if raw.ndim != 3 or raw.shape[1] != N_JOINTS or raw.shape[2] != 3:
        raise ValueError(f"Expected raw shape (T,{N_JOINTS},3), got {raw.shape}")
    if len(raw) < MIN_FRAMES:
        raise ValueError(f"Too short: {len(raw)} frames")
    for j in range(N_JOINTS):
        for c in range(3):
            raw[:,j,c] = _smooth(raw[:,j,c])
    raw = _fill_missing(raw)
    out = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
    for j in range(N_JOINTS):
        for c in range(3):
            out[:,j,c] = _resample(raw[:,j,c], TARGET_LEN)
    out = _normalise(out)
    return safe_np(out, clip=5.0)

# â”€â”€ scalar features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def extract_scalars(kps: np.ndarray) -> np.ndarray:
    kps = safe_np(kps, clip=5.0)

    def _ang(a, b, c):
        ba = a - b;  bc = c - b
        denom = np.maximum(
            np.linalg.norm(ba, axis=1) * np.linalg.norm(bc, axis=1),
            EPS)
        cos = np.sum(ba*bc, axis=1) / denom
        return np.degrees(np.arccos(np.clip(cos, -1, 1)))

    lk = _ang(kps[:,0,:], kps[:,1,:], kps[:,2,:])
    rk = _ang(kps[:,3,:], kps[:,4,:], kps[:,5,:])

    lr  = float(np.nanmax(lk) - np.nanmin(lk))
    rr  = float(np.nanmax(rk) - np.nanmin(rk))
    sym = abs(lr-rr) / ((lr+rr)/2.0+1e-6) * 100.0 if (lr+rr) > 1.0 else 0.0

    vel_arr = np.abs(np.diff(lk)) * FPS
    vel     = float(np.max(vel_arr)) if len(vel_arr) > 0 else 0.0

    hip  = (kps[:,0,:] + kps[:,3,:]) / 2.0
    jerk_arr = np.diff(np.diff(hip, axis=0), axis=0)
    jerk = float(np.nanmean(np.linalg.norm(jerk_arr, axis=1))) if len(jerk_arr) else 0.0

    peak  = int(np.argmin(lk))
    trunk = float(abs(hip[peak, 0] - hip[0, 0]))

    lk_c = lk - np.nanmean(lk);  rk_c = rk - np.nanmean(rk)
    if np.nanstd(lk_c) > 1e-6 and np.nanstd(rk_c) > 1e-6:
        lag = float(abs(
            np.argmax(np.correlate(lk_c, rk_c, 'full')) - (len(lk)-1)))
    else:
        lag = 0.0

    diff_std = float(np.nanstd(np.diff(lk)))
    smooth   = float(np.clip(1.0 / (diff_std + 1e-6) / 100.0, 0.0, 1.0))

    scalars = np.array([
        np.clip(lr   / 180.0, 0, 1),
        np.clip(rr   / 180.0, 0, 1),
        np.clip(np.min(lk) / 180.0, 0, 1),
        np.clip(np.min(rk) / 180.0, 0, 1),
        np.clip(sym  / 100.0, 0, 5),
        np.clip(vel  / 200.0, 0, 5),
        np.clip(jerk / 10.0,  0, 5),
        np.clip(trunk,        0, 5),
        np.clip(lag  / TARGET_LEN, 0, 1),
        smooth,
    ], dtype=np.float32)

    return safe_np(scalars, fill=0.0, clip=10.0)

# â”€â”€ PS1 interface â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def validate_ps1_frame(frame: dict) -> bool:
    for j in ['HipLeft','KneeLeft','AnkleLeft',
              'HipRight','KneeRight','AnkleRight']:
        for ax in ['x','y','z']:
            v = frame.get(f'{j}_{ax}')
            if v is None or np.isnan(float(v)) or abs(float(v)) > 10.0:
                return False
    return True

def ps1_frame_to_keypoint(frame: dict) -> np.ndarray:
    row = np.zeros((N_JOINTS, 3), dtype=np.float32)
    for ji, jname in enumerate(['HipLeft','KneeLeft','AnkleLeft',
                                  'HipRight','KneeRight','AnkleRight']):
        for ci, ax in enumerate(['x','y','z']):
            row[ji, ci] = float(frame.get(f'{jname}_{ax}', 0.0))
    return row

# â”€â”€ dataset loaders â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def _kps_from_df_rows(df, xyz_cols):
    return df[xyz_cols].values.astype(np.float32).reshape(len(df), N_JOINTS, 3)

def load_uiprmd() -> List[dict]:
    if not os.path.exists(UIPRMD_CSV):
        print(f"UI-PRMD CSV not found: {UIPRMD_CSV}"); return []
    df       = pd.read_csv(UIPRMD_CSV)
    xyz_cols = [f'{j}_{ax}' for j in JOINT_ORDER for ax in ['x','y','z']]
    samples  = []
    for (subj, mov, ex_id, label_val), grp in df.groupby(
            ['subject','movement','exercise_id','label']):
        grp   = grp.sort_values('frame')
        label = int(label_val)
        raw   = _kps_from_df_rows(grp, xyz_cols) / 1000.0
        try:
            kps = process_keypoints(raw)
            samples.append({
                'keypoints': kps, 'adj': ADJ,
                'label':    label,
                'quality':  float(label),
                'exercise': grp['exercise'].iloc[0],
                'subject':  f'uiprmd_{subj}',
                'source':   'uiprmd',
                'scalars':  extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip uiprmd {subj}/{mov}/{ex_id}: {e}")
    print(f"UI-PRMD: {len(samples)} samples")
    return samples

def load_kimore() -> List[dict]:
    if not os.path.exists(KIMORE_CSV):
        print(f"KIMORE not found: {KIMORE_CSV}"); return []
    df = pd.read_csv(KIMORE_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    df['quality_norm'] = df[['cTS','cPO','cCF']].mean(axis=1) / 50.0
    samples = []
    for subj, grp in df.groupby('Subject'):
        grp     = grp.sort_values('FrameID')
        label   = int(grp['label'].iloc[0])
        quality = float(grp['quality_norm'].mean())
        raw     = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps = process_keypoints(raw)
            samples.append({
                'keypoints': kps, 'adj': ADJ,
                'label':    label,
                'quality':  quality,
                'exercise': 'squat',
                'subject':  f'kimore_{subj}',
                'source':   'kimore',
                'scalars':  extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip kimore {subj}: {e}")
    print(f"KIMORE: {len(samples)} samples")
    return samples

def load_keraal() -> List[dict]:
    if not os.path.exists(KERAAL_CSV):
        print(f"Keraal not found: {KERAAL_CSV}"); return []
    df = pd.read_csv(KERAAL_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    samples = []
    for ann_key, grp in df.groupby('ann_key'):
        grp   = grp.sort_values('FrameID') if 'FrameID' in grp.columns else grp
        label = int(grp['label'].iloc[0])
        raw   = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps  = process_keypoints(raw)
            subj = '-'.join(str(ann_key).split('-')[:3])
            samples.append({
                'keypoints': kps, 'adj': ADJ,
                'label':    label,
                'quality':  float(label),
                'exercise': 'ctk_squat',
                'subject':  f'keraal_{subj}',
                'source':   'keraal',
                'scalars':  extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip keraal {ann_key}: {e}")
    print(f"Keraal: {len(samples)} samples")
    return samples

def build_master_dataset() -> List[dict]:
    samples = load_uiprmd() + load_kimore() + load_keraal()
    clean   = []
    dropped_by_source = {}
    for s in samples:
        source = s.get('source', 'unknown')
        try:
            kps = safe_np(s['keypoints'], clip=5.0)
            scalars = safe_np(s.get('scalars', extract_scalars(kps)), clip=10.0)
            quality = float(np.nan_to_num(s.get('quality', s.get('label', 1)),
                                          nan=float(s.get('label', 1))))

            if kps.shape != (TARGET_LEN, N_JOINTS, 3):
                raise ValueError(f"bad keypoint shape {kps.shape}")
            if scalars.shape != (10,):
                raise ValueError(f"bad scalar shape {scalars.shape}")
            if not valid_np(kps) or not valid_np(scalars):
                raise ValueError("NaN/Inf after sanitizing")

            s['keypoints'] = kps
            s['scalars'] = scalars
            s['label'] = int(s.get('label', 1))
            s['quality'] = float(np.clip(quality, 0.0, 1.0))
            clean.append(s)
        except Exception as e:
            print(f"  DROP corrupt: {s.get('subject', 'unknown')} "
                  f"{s.get('exercise', 'unknown')} ({e})")
            dropped_by_source[source] = dropped_by_source.get(source, 0) + 1
            continue
    print(f"\nMaster: {len(clean)} samples "
          f"(dropped {len(samples)-len(clean)} corrupt)")
    print(f"  correct={sum(s['label']==1 for s in clean)}"
          f"  incorrect={sum(s['label']==0 for s in clean)}")
    print(f"  sources:   {set(s['source']   for s in clean)}")
    print(f"  exercises: {set(s['exercise'] for s in clean)}")
    if dropped_by_source:
        print(f"  dropped_by_source: {dropped_by_source}")
    return clean

# â”€â”€ LOSO helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def loso_split(samples, held_out):
    return ([s for s in samples if s['subject'] != held_out],
            [s for s in samples if s['subject'] == held_out])

def get_uiprmd_subjects(samples):
    return sorted({s['subject'] for s in samples if s['source'] == 'uiprmd'})

# â”€â”€ PyTorch dataset â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class RehabDataset(Dataset):
    def __init__(self, samples, augment=False):
        self.samples = samples
        self.augment = augment

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        s   = self.samples[idx]
        kps = safe_np(s['keypoints'], clip=5.0)
        if self.augment:
            kps = self._aug(kps)
        kps = safe_np(kps, clip=5.0)
        scalars = safe_np(s['scalars'], clip=10.0)
        return {
            'keypoints': torch.from_numpy(kps).permute(2,0,1).float(),
            'adj':       torch.from_numpy(ADJ).float(),
            'label':     torch.tensor(s['label'],   dtype=torch.long),
            'quality':   torch.tensor(s['quality'], dtype=torch.float32),
            'scalars':   torch.from_numpy(scalars).float(),
            'exercise':  s['exercise'],
            'subject':   s['subject'],
        }

    def _aug(self, kps):
        kps = safe_np(kps, clip=5.0)
        if np.random.rand() < 0.5:
            n = max(int(TARGET_LEN * np.random.uniform(0.8, 1.2)), MIN_FRAMES)
            tmp = np.zeros((n, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    tmp[:,j,c] = _resample(kps[:,j,c], n)
            kps = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    kps[:,j,c] = _resample(tmp[:,j,c], TARGET_LEN)
        if np.random.rand() < 0.5:
            kps[:,[0,1,2,3,4,5],:] = kps[:,[3,4,5,0,1,2],:]
            kps[:,:,0] *= -1.0
        if np.random.rand() < 0.5:
            kps += np.random.randn(*kps.shape).astype(np.float32) * 0.02
        if np.random.rand() < 0.3:
            th  = np.radians(np.random.uniform(-10, 10))
            x   = kps[:,:,0].copy()
            z   = kps[:,:,2].copy()
            kps[:,:,0] = np.cos(th)*x - np.sin(th)*z
            kps[:,:,2] = np.sin(th)*x + np.cos(th)*z
        return safe_np(kps, clip=5.0)

def collate(batch):
    return {k: torch.stack([b[k] for b in batch])
              if isinstance(batch[0][k], torch.Tensor)
              else [b[k] for b in batch]
            for k in batch[0]}

def build_loaders(train, test, batch_size=32):
    from collections import Counter
    from torch.utils.data import WeightedRandomSampler
    ex_counts = Counter(s['exercise'] for s in train)
    total     = len(train)
    weights   = [total / ex_counts[s['exercise']] for s in train]
    sampler   = WeightedRandomSampler(weights, len(train), replacement=True)
    return (
        DataLoader(RehabDataset(train, True),  batch_size,
                   sampler=sampler, collate_fn=collate, num_workers=2),
        DataLoader(RehabDataset(test,  False), batch_size, False,
                   collate_fn=collate, num_workers=2),
    )

# â”€â”€ ST-GCN â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class STGCNBlock(nn.Module):
    def __init__(self, c_in, c_out, ks=9, stride=1):
        super().__init__()
        self.gcn = nn.Linear(c_in, c_out)
        self.tcn = nn.Sequential(
            nn.BatchNorm2d(c_out), nn.ReLU(),
            nn.Conv2d(c_out, c_out, (ks,1), (stride,1), ((ks-1)//2, 0)),
            nn.BatchNorm2d(c_out), nn.Dropout(0.1))
        self.res = (nn.Sequential(nn.Conv2d(c_in, c_out, 1, (stride,1)),
                                   nn.BatchNorm2d(c_out))
                    if c_in != c_out or stride != 1 else nn.Identity())
        self.relu = nn.ReLU()

    def forward(self, x, A):
        B, C, T, V = x.shape
        if A.dim() == 3: A = A[0]
        xs      = x.permute(0,2,3,1).contiguous().view(B*T, V, C)
        xs      = torch.bmm(A.unsqueeze(0).expand(B*T,-1,-1), xs)
        xs      = self.gcn(xs).view(B, T, V, -1).permute(0,3,1,2)
        tcn_out = self.tcn(xs)
        res_out = self.res(x)
        # guard NaN in residual stream
        if torch.isnan(tcn_out).any() or torch.isnan(res_out).any():
            tcn_out = torch.nan_to_num(tcn_out, 0.0)
            res_out = torch.nan_to_num(res_out, 0.0)
        return self.relu(tcn_out + res_out)

class STGCN(nn.Module):
    def __init__(self):
        super().__init__()
        # LayerNorm instead of BatchNorm1d â€” stable across mixed-source batches
        self.input_norm = nn.LayerNorm(IN_CHANNELS * N_JOINTS)
        self.b1 = STGCNBlock(3,  32)
        self.b2 = STGCNBlock(32, 64)
        self.b3 = STGCNBlock(64, 128)

    def forward(self, x, A):
        B, C, T, V = x.shape
        x  = torch.clamp(x, -5.0, 5.0)          # hard clamp before norm
        xf = x.permute(0,2,1,3).contiguous().view(B*T, C*V)
        xf = self.input_norm(xf).view(B, T, C, V).permute(0,2,1,3)
        return self.b3(self.b2(self.b1(xf, A), A), A).mean(3).permute(0,2,1)

# â”€â”€ clinical transformer â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class ClinicalTransformer(nn.Module):
    def __init__(self, d=128, heads=4, d_ff=256, drop=0.1):
        super().__init__()
        pe  = torch.zeros(TARGET_LEN+1, d)
        pos = torch.arange(TARGET_LEN+1).unsqueeze(1).float()
        div = torch.exp(torch.arange(0,d,2).float() * (-np.log(10000.0)/d))
        pe[:,0::2] = torch.sin(pos*div)
        pe[:,1::2] = torch.cos(pos*div)
        self.register_buffer('pe', pe.unsqueeze(0))
        self.cls  = nn.Parameter(torch.randn(1,1,d) * 0.02)
        self.drop = nn.Dropout(drop)
        mk = lambda: nn.TransformerEncoderLayer(
            d, heads, d_ff, drop, batch_first=True, norm_first=True)
        self.l1, self.l2 = mk(), mk()
        self.n1, self.n2 = nn.LayerNorm(d), nn.LayerNorm(d)

    def forward(self, x, return_attn=False):
        B, T, _ = x.shape
        x = torch.cat([self.cls.expand(B,-1,-1), x], 1)
        x = self.drop(x + self.pe[:, :T+1, :])
        x = self.n1(self.l1(x))
        if return_attn:
            sa  = self.l2.self_attn
            xn  = self.l2.norm1(x)
            out, attn = sa(xn, xn, xn,
                           need_weights=True, average_attn_weights=True)
            x2  = x + self.l2.dropout1(out)
            xn2 = self.l2.norm2(x2)
            ff  = self.l2.linear2(self.l2.dropout(
                      self.l2.activation(self.l2.linear1(xn2))))
            return self.n2(x2 + self.l2.dropout2(ff))[:,0,:], attn
        return self.n2(self.l2(x))[:,0,:], None

# â”€â”€ full model â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class RehabNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.stgcn    = STGCN()
        self.trans    = ClinicalTransformer()
        self.mode_emb = nn.Embedding(N_MODES, 16)
        self.shared   = nn.Sequential(
            nn.Linear(128+10+16, 128), nn.LayerNorm(128),
            nn.ReLU(), nn.Dropout(0.2))
        self.classify_head = nn.Linear(128, 2)
        self.quality_head  = nn.Linear(128, 1)
        # xavier init
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def encode(self, kps, adj, return_attn=False):
        if adj.dim() == 3: adj = adj[0]
        return self.trans(self.stgcn(kps, adj), return_attn)

    def forward(self, kps, adj, scalars, mode=None):
        B = kps.shape[0]
        if mode is None:
            mode = torch.zeros(B, dtype=torch.long, device=kps.device)
        cls, _    = self.encode(kps, adj)
        mode_feat = self.mode_emb(mode)
        x         = self.shared(torch.cat([cls, scalars, mode_feat], 1))
        logits    = self.classify_head(x)
        quality   = self.quality_head(x).squeeze(1)
        return logits, quality

# â”€â”€ training â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def train_epoch(model, loader, opt, device, epoch=0):
    model.train()
    A = torch.from_numpy(ADJ).to(device)
    total = 0.0
    used_batches = 0
    skipped_batches = 0
    dropped_samples = 0

    for batch_idx, b in enumerate(loader):
        kps = b['keypoints'].to(device)
        lbl = b['label'].to(device)
        qual = b['quality'].to(device)
        sc = b['scalars'].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1) &
            torch.isfinite(sc).all(1) &
            torch.isfinite(qual)
        )

        if not good.all():
            dropped_samples += int((~good).sum().item())
            kps = kps[good]
            lbl = lbl[good]
            qual = qual[good]
            sc = sc[good]

        if kps.shape[0] == 0:
            skipped_batches += 1
            continue

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        logits, qpred = model(kps, A, sc, mode)

        if not torch.isfinite(logits).all() or not torch.isfinite(qpred).all():
            if skipped_batches < 3:
                print(f"  [ep{epoch}] unstable output batch {batch_idx} - skip")
            skipped_batches += 1
            continue

        cls_loss = F.cross_entropy(logits, lbl)
        qual = torch.clamp(qual, 0.0, 1.0)
        qual_loss = F.mse_loss(torch.sigmoid(qpred), qual)
        loss = cls_loss + 0.3 * qual_loss

        if not torch.isfinite(loss):
            skipped_batches += 1
            continue

        opt.zero_grad(set_to_none=True)
        loss.backward()

        if epoch == 0 and batch_idx == 0:
            total_norm = sum(p.grad.norm()**2
                            for p in model.parameters()
                            if p.grad is not None) ** 0.5
            print(f"  [ep0 batch0] grad norm = {total_norm:.3f}")

        nn.utils.clip_grad_norm_(model.parameters(), 0.5)
        opt.step()
        total += loss.item()
        used_batches += 1

    if dropped_samples > 0 or skipped_batches > 0:
        print(f"  epoch {epoch}: dropped {dropped_samples} bad samples, "
              f"skipped {skipped_batches} empty/unstable batches")
    return total / max(used_batches, 1)
# evaluation

@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    A = torch.from_numpy(ADJ).to(device)
    preds, labels, quals, qpreds = [], [], [], []

    for b in loader:
        kps = b['keypoints'].to(device)
        sc = b['scalars'].to(device)
        qual = b['quality'].to(device)
        lbl = b['label'].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1) &
            torch.isfinite(sc).all(1) &
            torch.isfinite(qual)
        )

        if good.sum().item() == 0:
            continue

        kps = kps[good]
        sc = sc[good]
        qual = qual[good]
        lbl = lbl[good]

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        lg, qp = model(kps, A, sc, mode)
        preds.extend(lg.argmax(1).cpu().numpy())
        labels.extend(lbl.cpu().numpy())
        quals.extend(qual.cpu().numpy())
        qpreds.extend(torch.sigmoid(qp).cpu().numpy())

    from sklearn.metrics import accuracy_score, f1_score, classification_report
    from scipy.stats import pearsonr

    if len(labels) == 0:
        print("  WARNING: no valid evaluation samples")
        return {'acc': 0.0, 'f1': 0.0, 'quality_r': 0.0}

    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='macro', zero_division=0)
    r = pearsonr(quals, qpreds)[0] if len(set(quals)) > 1 else 0.0
    print(classification_report(labels, preds,
          target_names=['incorrect','correct'], zero_division=0))
    print(f"  quality Pearson r={r:.3f}")
    return {'acc': acc, 'f1': f1, 'quality_r': r}
# LOSO training loop â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def run_loso(n_epochs=40, batch_size=32, lr=3e-4):
    device   = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Device: {device}")
    samples  = build_master_dataset()
    subjects = get_uiprmd_subjects(samples)
    results  = []

    for subj in subjects:
        print(f"\n{'='*55}")
        print(f"LOSO held-out: {subj}")
        print(f"{'='*55}")
        tr, te = loso_split(samples, subj)
        tl, vl = build_loaders(tr, te, batch_size)

        model = RehabNet().to(device)
        opt   = torch.optim.AdamW(model.parameters(),
                                  lr=lr, weight_decay=1e-4)

        def lr_lambda(ep):
            warmup = 5
            if ep < warmup:
                return (ep + 1) / warmup
            progress = (ep - warmup) / max(n_epochs - warmup, 1)
            return 0.5 * (1 + np.cos(np.pi * progress))

        sch = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda)

        for ep in range(n_epochs):
            loss = train_epoch(model, tl, opt, device, epoch=ep)
            sch.step()
            if (ep+1) % 5 == 0:
                print(f"  ep{ep+1:3d} | loss={loss:.4f} | "
                      f"lr={sch.get_last_lr()[0]:.2e}")

        print(f"\nâ”€â”€ Evaluation: {subj} â”€â”€")
        results.append(evaluate(model, vl, device))

    accs = [r['acc'] for r in results]
    f1s  = [r['f1']  for r in results]
    print(f"\n{'='*55}")
    print(f"LOSO FINAL RESULTS")
    print(f"  Accuracy : {np.mean(accs):.3f} Â± {np.std(accs):.3f}")
    print(f"  F1 macro : {np.mean(f1s):.3f}  Â± {np.std(f1s):.3f}")
    print(f"{'='*55}")
    return results

# â”€â”€ live inference â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class LiveInference:
    def __init__(self, model: RehabNet, device='cpu', active_mode=0):
        self.model       = model.eval().to(device)
        self.device      = torch.device(device)
        self.buffer: List[np.ndarray] = []
        self.active_mode = active_mode
        self.last_result = None

    def set_mode(self, mode: int):
        assert mode in HARDWARE_MODES, f"mode must be 0-4, got {mode}"
        self.active_mode = mode
        print(f"Hardware mode â†’ {HARDWARE_MODES[mode]['name']}")

    def push_frame(self, ps1_frame: dict) -> Optional[dict]:
        if not validate_ps1_frame(ps1_frame):
            return None
        self.buffer.append(ps1_frame_to_keypoint(ps1_frame))
        if len(self.buffer) > TARGET_LEN:
            self.buffer = self.buffer[-TARGET_LEN:]
        if len(self.buffer) < TARGET_LEN:
            return None
        return self._infer()

    @torch.no_grad()
    def _infer(self) -> dict:
        raw  = np.stack(self.buffer, axis=0)
        kps  = process_keypoints(raw.copy())
        sc   = extract_scalars(kps)
        vel_proxy = float(sc[5])

        kps_t  = torch.from_numpy(kps).permute(2,0,1).unsqueeze(0).to(self.device)
        sc_t   = torch.from_numpy(sc).unsqueeze(0).to(self.device)
        A      = torch.from_numpy(ADJ).to(self.device)
        mode_t = torch.tensor([self.active_mode],
                               dtype=torch.long, device=self.device)

        logits, quality = self.model(kps_t, A, sc_t, mode_t)
        prob    = F.softmax(logits, dim=1)[0]
        label   = int(logits.argmax(1).item())
        q       = float(torch.sigmoid(quality).item())
        torque  = compute_torque(q, self.active_mode, vel_proxy)

        self.last_result = {
            'label':         label,
            'correct_prob':  float(prob[1]),
            'quality_score': q,
            'torque_signal': torque,
            'mode':          self.active_mode,
            'mode_name':     HARDWARE_MODES[self.active_mode]['name'],
            'error_flag':    label == 0,
        }
        return self.last_result

# â”€â”€ smoke test â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

if __name__ == '__main__':
    print("â”€â”€ smoke test â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    B   = 4
    kps = torch.randn(B, 3, TARGET_LEN, N_JOINTS)
    adj = torch.from_numpy(ADJ).unsqueeze(0).expand(B,-1,-1)
    sc  = torch.rand(B, 10)

    m = RehabNet()
    for mode_id in range(N_MODES):
        mode = torch.full((B,), mode_id, dtype=torch.long)
        lg, q = m(kps, adj, sc, mode)
        print(f"  mode={mode_id} ({HARDWARE_MODES[mode_id]['name']:22s})"
              f"  logits={lg.shape}  quality={q.shape}")

    cls_vec, attn = m.encode(kps, adj, return_attn=True)
    print(f"  CLS={cls_vec.shape}  attn={attn.shape}")

    live = LiveInference(m, active_mode=2)
    for _ in range(TARGET_LEN + 5):
        frame = {f'{j}_{ax}': float(np.random.randn() * 0.3)
                 for j in ['HipLeft','KneeLeft','AnkleLeft',
                            'HipRight','KneeRight','AnkleRight']
                 for ax in ['x','y','z']}
        out = live.push_frame(frame)
    print(f"  live result: {out}")
    live.set_mode(4)
    print("â”€â”€ passed â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
```

```python
"""
ST-GCN + Transformer + BiLSTM fusion model.

Run your cleaned transformer/data code first, then run this file.

This file reuses from your transformer code:
  STGCN, ClinicalTransformer, ADJ, N_MODES
  build_master_dataset, get_uiprmd_subjects, loso_split, build_loaders

Architecture:
  keypoints -> ST-GCN -> frame features (B, 150, 128)
                  |-> ClinicalTransformer -> transformer embedding (B, 128)
                  |-> BiLSTM + attention -> recurrent embedding (B, 128)
  scalars + mode embedding are concatenated with both embeddings.
"""

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F


class LSTMAttentionPool(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.score = nn.Sequential(
            nn.Linear(dim, dim // 2),
            nn.Tanh(),
            nn.Linear(dim // 2, 1),
        )

    def forward(self, x):
        # x: (B, T, D)
        w = self.score(x).squeeze(-1)
        w = torch.softmax(w, dim=1)
        pooled = torch.sum(x * w.unsqueeze(-1), dim=1)
        return pooled, w


class STGCNTransformerBiLSTMFusion(nn.Module):
    def __init__(
        self,
        lstm_hidden=64,
        lstm_layers=1,
        dropout=0.25,
        n_classes=2,
    ):
        super().__init__()

        self.stgcn = STGCN()
        self.trans = ClinicalTransformer()

        self.lstm = nn.LSTM(
            input_size=128,
            hidden_size=lstm_hidden,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if lstm_layers > 1 else 0.0,
        )

        self.lstm_dim = lstm_hidden * 2
        self.lstm_attn = LSTMAttentionPool(self.lstm_dim)
        self.mode_emb = nn.Embedding(N_MODES, 16)

        fusion_dim = 128 + self.lstm_dim + 10 + 16
        self.shared = nn.Sequential(
            nn.Linear(fusion_dim, 160),
            nn.LayerNorm(160),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(160, 96),
            nn.LayerNorm(96),
            nn.ReLU(),
            nn.Dropout(dropout),
        )

        self.classify_head = nn.Linear(96, n_classes)
        self.quality_head = nn.Linear(96, 1)

        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def encode(self, kps, adj, return_attn=False):
        if adj.dim() == 3:
            adj = adj[0]

        frame_feat = self.stgcn(kps, adj)  # (B, 150, 128)
        trans_vec, trans_attn = self.trans(frame_feat, return_attn=return_attn)

        lstm_out, _ = self.lstm(frame_feat)
        lstm_vec, lstm_attn = self.lstm_attn(lstm_out)

        if return_attn:
            return trans_vec, lstm_vec, {
                "transformer_attn": trans_attn,
                "bilstm_attn": lstm_attn,
            }

        return trans_vec, lstm_vec, None

    def forward(self, kps, adj, scalars, mode=None):
        b = kps.shape[0]
        if mode is None:
            mode = torch.zeros(b, dtype=torch.long, device=kps.device)

        scalars = torch.nan_to_num(scalars, nan=0.0, posinf=10.0, neginf=0.0)
        scalars = torch.clamp(scalars, -10.0, 10.0)

        trans_vec, lstm_vec, _ = self.encode(kps, adj, return_attn=False)
        mode_feat = self.mode_emb(mode)

        fused = torch.cat([trans_vec, lstm_vec, scalars, mode_feat], dim=1)
        fused = torch.nan_to_num(fused, nan=0.0, posinf=10.0, neginf=-10.0)

        x = self.shared(fused)
        logits = self.classify_head(x)
        quality = self.quality_head(x).squeeze(1)
        return logits, quality


def train_epoch_fusion(model, loader, opt, device, epoch=0):
    model.train()
    A = torch.from_numpy(ADJ).to(device)
    total = 0.0
    used_batches = 0
    dropped_samples = 0
    skipped_batches = 0

    for batch_idx, b in enumerate(loader):
        kps = b["keypoints"].to(device)
        lbl = b["label"].to(device)
        qual = b["quality"].to(device)
        sc = b["scalars"].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1)
            & torch.isfinite(sc).all(1)
            & torch.isfinite(qual)
        )

        if not good.all():
            dropped_samples += int((~good).sum().item())
            kps = kps[good]
            lbl = lbl[good]
            qual = qual[good]
            sc = sc[good]

        if kps.shape[0] == 0:
            skipped_batches += 1
            continue

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        logits, qpred = model(kps, A, sc, mode)

        if not torch.isfinite(logits).all() or not torch.isfinite(qpred).all():
            if skipped_batches < 3:
                print(f"  [fusion ep{epoch}] unstable output batch {batch_idx} - skip")
            skipped_batches += 1
            continue

        cls_loss = F.cross_entropy(logits, lbl)
        qual = torch.clamp(qual, 0.0, 1.0)
        qual_loss = F.mse_loss(torch.sigmoid(qpred), qual)
        loss = cls_loss + 0.3 * qual_loss

        if not torch.isfinite(loss):
            skipped_batches += 1
            continue

        opt.zero_grad(set_to_none=True)
        loss.backward()

        if epoch == 0 and batch_idx == 0:
            total_norm = sum(
                p.grad.norm() ** 2
                for p in model.parameters()
                if p.grad is not None
            ) ** 0.5
            print(f"  [Fusion ep0 batch0] grad norm = {total_norm:.3f}")

        nn.utils.clip_grad_norm_(model.parameters(), 0.75)
        opt.step()

        total += loss.item()
        used_batches += 1

    if dropped_samples > 0 or skipped_batches > 0:
        print(
            f"  epoch {epoch}: dropped {dropped_samples} bad samples, "
            f"skipped {skipped_batches} empty/unstable batches"
        )

    return total / max(used_batches, 1)


@torch.no_grad()
def evaluate_fusion(model, loader, device):
    model.eval()
    A = torch.from_numpy(ADJ).to(device)
    preds, labels, quals, qpreds = [], [], [], []

    for b in loader:
        kps = b["keypoints"].to(device)
        sc = b["scalars"].to(device)
        qual = b["quality"].to(device)
        lbl = b["label"].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1)
            & torch.isfinite(sc).all(1)
            & torch.isfinite(qual)
        )

        if good.sum().item() == 0:
            continue

        kps = kps[good]
        sc = sc[good]
        qual = qual[good]
        lbl = lbl[good]

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        logits, qp = model(kps, A, sc, mode)

        preds.extend(logits.argmax(1).cpu().numpy())
        labels.extend(lbl.cpu().numpy())
        quals.extend(qual.cpu().numpy())
        qpreds.extend(torch.sigmoid(qp).cpu().numpy())

    from sklearn.metrics import accuracy_score, f1_score, classification_report
    from scipy.stats import pearsonr

    if len(labels) == 0:
        print("  WARNING: no valid evaluation samples")
        return {"acc": 0.0, "f1": 0.0, "quality_r": 0.0}

    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average="macro", zero_division=0)
    r = pearsonr(quals, qpreds)[0] if len(set(quals)) > 1 else 0.0

    print(
        classification_report(
            labels,
            preds,
            target_names=["incorrect", "correct"],
            zero_division=0,
        )
    )
    print(f"  quality Pearson r={r:.3f}")

    return {"acc": acc, "f1": f1, "quality_r": r}


def run_loso_fusion(n_epochs=30, batch_size=32, lr=2e-4):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Device: {device}")

    samples = build_master_dataset()
    subjects = get_uiprmd_subjects(samples)
    results = []

    for subj in subjects:
        print(f"\n{'=' * 55}")
        print(f"Fusion LOSO held-out: {subj}")
        print(f"{'=' * 55}")

        train, test = loso_split(samples, subj)
        train_loader, test_loader = build_loaders(train, test, batch_size)

        model = STGCNTransformerBiLSTMFusion(
            lstm_hidden=64,
            lstm_layers=1,
            dropout=0.25,
        ).to(device)

        opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)

        def lr_lambda(ep):
            warmup = 5
            if ep < warmup:
                return (ep + 1) / warmup
            progress = (ep - warmup) / max(n_epochs - warmup, 1)
            return 0.5 * (1 + np.cos(np.pi * progress))

        sch = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda)

        for ep in range(n_epochs):
            loss = train_epoch_fusion(model, train_loader, opt, device, epoch=ep)
            sch.step()
            if (ep + 1) % 5 == 0:
                print(
                    f"  ep{ep + 1:3d} | loss={loss:.4f} | "
                    f"lr={sch.get_last_lr()[0]:.2e}"
                )

        print(f"\nEvaluation: {subj}")
        results.append(evaluate_fusion(model, test_loader, device))

    accs = [r["acc"] for r in results]
    f1s = [r["f1"] for r in results]
    qrs = [r["quality_r"] for r in results]

    print(f"\n{'=' * 55}")
    print("FUSION LOSO FINAL RESULTS")
    print(f"  Accuracy  : {np.mean(accs):.3f} +/- {np.std(accs):.3f}")
    print(f"  F1 macro  : {np.mean(f1s):.3f} +/- {np.std(f1s):.3f}")
    print(f"  Quality r : {np.mean(qrs):.3f} +/- {np.std(qrs):.3f}")
    print(f"{'=' * 55}")

    return results
```

```python
import os
import copy
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d
from typing import Dict, List, Optional, Tuple
from collections import Counter

# ── config ────────────────────────────────────────────────────────────────────

TARGET_LEN  = 150
N_JOINTS    = 6
IN_CHANNELS = 3
MIN_FRAMES  = 20
FPS         = 30
N_EPOCHS    = 60   # increased from 30

UIPRMD_CSV = '/content/drive/MyDrive/datasets/UIPRMD_timeseries.csv'
KIMORE_CSV = '/content/drive/MyDrive/datasets/squats_tabular_timeseries_binary.csv'
KERAAL_CSV = '/content/drive/MyDrive/datasets/keraal_final_binary.csv'

JOINT_ORDER = ['l_hip','l_knee','l_ankle','r_hip','r_knee','r_ankle']

EXERCISE_LIST = [
    'deep_squat','hurdle_step','inline_lunge','side_lunge',
    'sit_to_stand','straight_leg_raise','squat','ctk_squat','unknown'
]
EXERCISE2IDX = {e: i for i, e in enumerate(EXERCISE_LIST)}
N_EXERCISES  = len(EXERCISE_LIST)

# ── hardware modes ────────────────────────────────────────────────────────────

HARDWARE_MODES = {
    0: {'name': 'free',              'scale': 0.0,  'bias': 0.0,  'min': 0.0, 'max': 0.0},
    1: {'name': 'constant_torque',   'scale': 0.0,  'bias': 0.5,  'min': 0.4, 'max': 0.6},
    2: {'name': 'resistive_torque',  'scale': 0.5,  'bias': 0.0,  'min': 0.0, 'max': 0.8},
    3: {'name': 'increasing_torque', 'scale': 1.0,  'bias': 0.0,  'min': 0.0, 'max': 1.0},
    4: {'name': 'dampening',         'scale': -0.5, 'bias': 0.5,  'min': 0.0, 'max': 0.5},
}
N_MODES = len(HARDWARE_MODES)

def compute_torque(quality: float, mode: int,
                   velocity_proxy: float = 0.0) -> float:
    m = HARDWARE_MODES.get(mode, HARDWARE_MODES[0])
    if mode == 4:
        torque = m['bias'] - abs(velocity_proxy) * 0.5
    else:
        torque = m['scale'] * quality + m['bias']
    return float(np.clip(torque, m['min'], m['max']))

# ── adjacency ─────────────────────────────────────────────────────────────────

def build_adjacency() -> np.ndarray:
    A = np.zeros((N_JOINTS, N_JOINTS), dtype=np.float32)
    for i, j in [(0,1),(1,2),(3,4),(4,5),(0,3),(1,4),(2,5)]:
        A[i,j] = A[j,i] = 1.0
    np.fill_diagonal(A, 1.0)
    d = np.diag(1.0 / np.sqrt(A.sum(axis=1) + 1e-6))
    return (d @ A @ d).astype(np.float32)

ADJ = build_adjacency()

# ── signal helpers ────────────────────────────────────────────────────────────

def _smooth(arr, win=7, poly=2):
    if len(arr) < win + 2:
        return arr
    w = win if win % 2 == 1 else win + 1
    w = max(w, poly+2 if (poly+2)%2==1 else poly+3)
    return savgol_filter(arr, w, poly)

def _resample(arr, n):
    if len(arr) == n:
        return arr
    return np.interp(np.linspace(0,1,n), np.linspace(0,1,len(arr)), arr)

def _fill_missing(kps):
    T, J, C = kps.shape
    for j in range(J):
        for c in range(C):
            col = kps[:,j,c]
            z   = col == 0.0
            if z.all():
                opp = j+3 if j < 3 else j-3
                kps[:,j,c] = kps[:,opp,c]
            elif z.any() and (~z).sum() >= 2:
                f = interp1d(np.where(~z)[0], col[~z],
                             bounds_error=False, fill_value='extrapolate')
                kps[:,j,c] = f(np.arange(T))
    return kps

def _normalise(kps):
    hip   = ((kps[:,0,:] + kps[:,3,:]) / 2.0).mean(axis=0)
    kps   = kps - hip
    l     = np.mean(np.linalg.norm(kps[:,0,:] - kps[:,2,:], axis=1))
    r     = np.mean(np.linalg.norm(kps[:,3,:] - kps[:,5,:], axis=1))
    scale = (l + r) / 2.0
    if scale < 1e-4:
        scale = 1.0
    kps   = kps / scale
    kps   = np.clip(kps, -5.0, 5.0)
    return kps.astype(np.float32)

def process_keypoints(raw: np.ndarray) -> np.ndarray:
    if len(raw) < MIN_FRAMES:
        raise ValueError(f"Too short: {len(raw)} frames")
    for j in range(N_JOINTS):
        for c in range(3):
            raw[:,j,c] = _smooth(raw[:,j,c])
    raw = _fill_missing(raw)
    out = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
    for j in range(N_JOINTS):
        for c in range(3):
            out[:,j,c] = _resample(raw[:,j,c], TARGET_LEN)
    out = _normalise(out)
    out = np.nan_to_num(out, nan=0.0, posinf=5.0, neginf=-5.0)
    return out

# ── scalar features ───────────────────────────────────────────────────────────

def extract_scalars(kps: np.ndarray) -> np.ndarray:
    def _ang(a, b, c):
        ba = a - b;  bc = c - b
        cos = np.sum(ba*bc, 1) / (
            np.linalg.norm(ba,1) * np.linalg.norm(bc,1) + 1e-6)
        return np.degrees(np.arccos(np.clip(cos, -1, 1)))

    lk = _ang(kps[:,0,:], kps[:,1,:], kps[:,2,:])
    rk = _ang(kps[:,3,:], kps[:,4,:], kps[:,5,:])

    lr  = float(np.max(lk) - np.min(lk))
    rr  = float(np.max(rk) - np.min(rk))
    sym = abs(lr-rr) / ((lr+rr)/2.0+1e-6) * 100.0 if (lr+rr) > 1.0 else 0.0

    vel_arr = np.abs(np.diff(lk)) * FPS
    vel     = float(np.max(vel_arr)) if len(vel_arr) > 0 else 0.0

    hip  = (kps[:,0,:] + kps[:,3,:]) / 2.0
    jerk = float(np.mean(np.linalg.norm(
               np.diff(np.diff(hip, axis=0), axis=0), axis=1)))

    peak  = int(np.argmin(lk))
    trunk = float(abs(hip[peak, 0] - hip[0, 0]))

    lk_c = lk - lk.mean();  rk_c = rk - rk.mean()
    if lk_c.std() > 1e-6 and rk_c.std() > 1e-6:
        lag = float(abs(
            np.argmax(np.correlate(lk_c, rk_c, 'full')) - (len(lk)-1)))
    else:
        lag = 0.0

    diff_std = float(np.std(np.diff(lk)))
    smooth   = float(np.clip(1.0 / (diff_std + 1e-6) / 100.0, 0.0, 1.0))

    scalars = np.array([
        np.clip(lr   / 180.0, 0, 1),
        np.clip(rr   / 180.0, 0, 1),
        np.clip(np.min(lk) / 180.0, 0, 1),
        np.clip(np.min(rk) / 180.0, 0, 1),
        np.clip(sym  / 100.0, 0, 5),
        np.clip(vel  / 200.0, 0, 5),
        np.clip(jerk / 10.0,  0, 5),
        np.clip(trunk,        0, 5),
        np.clip(lag  / TARGET_LEN, 0, 1),
        smooth,
    ], dtype=np.float32)

    return np.nan_to_num(scalars, nan=0.0, posinf=1.0, neginf=0.0)

# ── PS1 interface ─────────────────────────────────────────────────────────────

def validate_ps1_frame(frame: dict) -> bool:
    for j in ['HipLeft','KneeLeft','AnkleLeft',
              'HipRight','KneeRight','AnkleRight']:
        for ax in ['x','y','z']:
            v = frame.get(f'{j}_{ax}')
            if v is None or np.isnan(float(v)) or abs(float(v)) > 10.0:
                return False
    return True

def ps1_frame_to_keypoint(frame: dict) -> np.ndarray:
    row = np.zeros((N_JOINTS, 3), dtype=np.float32)
    for ji, jname in enumerate(['HipLeft','KneeLeft','AnkleLeft',
                                  'HipRight','KneeRight','AnkleRight']):
        for ci, ax in enumerate(['x','y','z']):
            row[ji, ci] = float(frame.get(f'{jname}_{ax}', 0.0))
    return row

# ── dataset loaders ───────────────────────────────────────────────────────────

def _kps_from_df_rows(df, xyz_cols):
    return df[xyz_cols].values.astype(np.float32).reshape(len(df), N_JOINTS, 3)

def load_uiprmd() -> List[dict]:
    if not os.path.exists(UIPRMD_CSV):
        print(f"UI-PRMD CSV not found: {UIPRMD_CSV}"); return []
    df       = pd.read_csv(UIPRMD_CSV)
    xyz_cols = [f'{j}_{ax}' for j in JOINT_ORDER for ax in ['x','y','z']]
    samples  = []
    for (subj, mov, ex_id, label_val), grp in df.groupby(
            ['subject','movement','exercise_id','label']):
        grp   = grp.sort_values('frame')
        label = int(label_val)
        raw   = _kps_from_df_rows(grp, xyz_cols) / 1000.0
        try:
            kps = process_keypoints(raw)
            ex  = grp['exercise'].iloc[0]
            samples.append({
                'keypoints':   kps,
                'adj':         ADJ,
                'label':       label,
                'quality':     float(label),
                'exercise':    ex,
                'exercise_idx': EXERCISE2IDX.get(ex, EXERCISE2IDX['unknown']),
                'subject':     f'uiprmd_{subj}',
                'source':      'uiprmd',
                'scalars':     extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip uiprmd {subj}/{mov}/{ex_id}: {e}")
    print(f"UI-PRMD: {len(samples)} samples")
    return samples

def load_kimore() -> List[dict]:
    if not os.path.exists(KIMORE_CSV):
        print(f"KIMORE not found: {KIMORE_CSV}"); return []
    df = pd.read_csv(KIMORE_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    df['quality_norm'] = df[['cTS','cPO','cCF']].mean(axis=1) / 50.0
    samples = []
    for subj, grp in df.groupby('Subject'):
        grp     = grp.sort_values('FrameID')
        label   = int(grp['label'].iloc[0])
        quality = float(grp['quality_norm'].mean())
        raw     = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps = process_keypoints(raw)
            samples.append({
                'keypoints':    kps,
                'adj':          ADJ,
                'label':        label,
                'quality':      quality,
                'exercise':     'squat',
                'exercise_idx': EXERCISE2IDX['squat'],
                'subject':      f'kimore_{subj}',
                'source':       'kimore',
                'scalars':      extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip kimore {subj}: {e}")
    print(f"KIMORE: {len(samples)} samples")
    return samples

def load_keraal() -> List[dict]:
    if not os.path.exists(KERAAL_CSV):
        print(f"Keraal not found: {KERAAL_CSV}"); return []
    df = pd.read_csv(KERAAL_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    samples = []
    for ann_key, grp in df.groupby('ann_key'):
        grp   = grp.sort_values('FrameID') if 'FrameID' in grp.columns else grp
        label = int(grp['label'].iloc[0])
        raw   = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps  = process_keypoints(raw)
            subj = '-'.join(str(ann_key).split('-')[:3])
            samples.append({
                'keypoints':    kps,
                'adj':          ADJ,
                'label':        label,
                'quality':      float(label),
                'exercise':     'ctk_squat',
                'exercise_idx': EXERCISE2IDX['ctk_squat'],
                'subject':      f'keraal_{subj}',
                'source':       'keraal',
                'scalars':      extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip keraal {ann_key}: {e}")
    print(f"Keraal: {len(samples)} samples")
    return samples

def build_master_dataset() -> List[dict]:
    samples = load_uiprmd() + load_kimore() + load_keraal()
    clean   = []
    for s in samples:
        if (np.isnan(s['keypoints']).any() or np.isinf(s['keypoints']).any() or
                np.isnan(s['scalars']).any()   or np.isinf(s['scalars']).any()):
            print(f"  DROP corrupt: {s['subject']} {s['exercise']}")
            continue
        clean.append(s)
    print(f"\nMaster: {len(clean)} samples "
          f"(dropped {len(samples)-len(clean)} corrupt)")
    print(f"  correct={sum(s['label']==1 for s in clean)}"
          f"  incorrect={sum(s['label']==0 for s in clean)}")
    print(f"  sources:   {set(s['source']   for s in clean)}")
    print(f"  exercises: {set(s['exercise'] for s in clean)}")
    return clean

# ── LOSO helpers ──────────────────────────────────────────────────────────────

def loso_split(samples, held_out):
    return ([s for s in samples if s['subject'] != held_out],
            [s for s in samples if s['subject'] == held_out])

def get_uiprmd_subjects(samples):
    return sorted({s['subject'] for s in samples if s['source'] == 'uiprmd'})

# ── PyTorch dataset ───────────────────────────────────────────────────────────

class RehabDataset(Dataset):
    def __init__(self, samples, augment=False):
        self.samples = samples
        self.augment = augment

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        s   = self.samples[idx]
        kps = s['keypoints'].copy()
        if self.augment:
            kps = self._aug(kps)
        return {
            'keypoints':    torch.from_numpy(kps).permute(2,0,1).float(),
            'adj':          torch.from_numpy(s['adj']).float(),
            'label':        torch.tensor(s['label'],        dtype=torch.long),
            'quality':      torch.tensor(s['quality'],      dtype=torch.float32),
            'scalars':      torch.from_numpy(s['scalars']).float(),
            'exercise_idx': torch.tensor(s['exercise_idx'], dtype=torch.long),
            'exercise':     s['exercise'],
            'subject':      s['subject'],
        }

    def _aug(self, kps):
        if np.random.rand() < 0.5:
            n = max(int(TARGET_LEN * np.random.uniform(0.8, 1.2)), MIN_FRAMES)
            tmp = np.zeros((n, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    tmp[:,j,c] = _resample(kps[:,j,c], n)
            kps = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    kps[:,j,c] = _resample(tmp[:,j,c], TARGET_LEN)
        if np.random.rand() < 0.5:
            kps[:,[0,1,2,3,4,5],:] = kps[:,[3,4,5,0,1,2],:]
            kps[:,:,0] *= -1.0
        if np.random.rand() < 0.5:
            kps += np.random.randn(*kps.shape).astype(np.float32) * 0.02
        if np.random.rand() < 0.3:
            th  = np.radians(np.random.uniform(-10, 10))
            x   = kps[:,:,0].copy()
            z   = kps[:,:,2].copy()
            kps[:,:,0] = np.cos(th)*x - np.sin(th)*z
            kps[:,:,2] = np.sin(th)*x + np.cos(th)*z
        return kps

def collate(batch):
    return {k: torch.stack([b[k] for b in batch])
              if isinstance(batch[0][k], torch.Tensor)
              else [b[k] for b in batch]
            for k in batch[0]}

def build_loaders(train_samples, test_samples, batch_size=32):
    from torch.utils.data import WeightedRandomSampler
    ex_counts = Counter(s['exercise'] for s in train_samples)
    total     = len(train_samples)
    weights   = [total / ex_counts[s['exercise']] for s in train_samples]
    sampler   = WeightedRandomSampler(weights, len(train_samples), replacement=True)
    return (
        DataLoader(RehabDataset(train_samples, True),  batch_size,
                   sampler=sampler, collate_fn=collate, num_workers=2),
        DataLoader(RehabDataset(test_samples,  False), batch_size, False,
                   collate_fn=collate, num_workers=2),
    )

# ── ST-GCN ────────────────────────────────────────────────────────────────────

class STGCNBlock(nn.Module):
    def __init__(self, c_in, c_out, ks=9, stride=1):
        super().__init__()
        self.gcn = nn.Linear(c_in, c_out)
        self.tcn = nn.Sequential(
            nn.BatchNorm2d(c_out), nn.ReLU(),
            nn.Conv2d(c_out, c_out, (ks,1), (stride,1), ((ks-1)//2, 0)),
            nn.BatchNorm2d(c_out), nn.Dropout(0.1))
        self.res = (nn.Sequential(nn.Conv2d(c_in, c_out, 1, (stride,1)),
                                   nn.BatchNorm2d(c_out))
                    if c_in != c_out or stride != 1 else nn.Identity())
        self.relu = nn.ReLU()

    def forward(self, x, A):
        B, C, T, V = x.shape
        if A.dim() == 3: A = A[0]
        xs      = x.permute(0,2,3,1).contiguous().view(B*T, V, C)
        xs      = torch.bmm(A.unsqueeze(0).expand(B*T,-1,-1), xs)
        xs      = self.gcn(xs).view(B, T, V, -1).permute(0,3,1,2)
        tcn_out = self.tcn(xs)
        res_out = self.res(x)
        if torch.isnan(tcn_out).any() or torch.isnan(res_out).any():
            tcn_out = torch.nan_to_num(tcn_out, 0.0)
            res_out = torch.nan_to_num(res_out, 0.0)
        return self.relu(tcn_out + res_out)

class STGCN(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_norm = nn.LayerNorm(IN_CHANNELS * N_JOINTS)
        self.b1 = STGCNBlock(3,  32)
        self.b2 = STGCNBlock(32, 64)
        self.b3 = STGCNBlock(64, 128)

    def forward(self, x, A):
        B, C, T, V = x.shape
        x  = torch.clamp(x, -5.0, 5.0)
        xf = x.permute(0,2,1,3).contiguous().view(B*T, C*V)
        xf = self.input_norm(xf).view(B, T, C, V).permute(0,2,1,3)
        return self.b3(self.b2(self.b1(xf, A), A), A).mean(3).permute(0,2,1)

# ── clinical transformer ──────────────────────────────────────────────────────

class ClinicalTransformer(nn.Module):
    def __init__(self, d=128, heads=4, d_ff=256, drop=0.1):
        super().__init__()
        pe  = torch.zeros(TARGET_LEN+1, d)
        pos = torch.arange(TARGET_LEN+1).unsqueeze(1).float()
        div = torch.exp(torch.arange(0,d,2).float() * (-np.log(10000.0)/d))
        pe[:,0::2] = torch.sin(pos*div)
        pe[:,1::2] = torch.cos(pos*div)
        self.register_buffer('pe', pe.unsqueeze(0))
        self.cls  = nn.Parameter(torch.randn(1,1,d) * 0.02)
        self.drop = nn.Dropout(drop)
        mk = lambda: nn.TransformerEncoderLayer(
            d, heads, d_ff, drop, batch_first=True, norm_first=True)
        self.l1, self.l2 = mk(), mk()
        self.n1, self.n2 = nn.LayerNorm(d), nn.LayerNorm(d)

    def forward(self, x, return_attn=False):
        B, T, _ = x.shape
        x = torch.cat([self.cls.expand(B,-1,-1), x], 1)
        x = self.drop(x + self.pe[:, :T+1, :])
        x = self.n1(self.l1(x))
        if return_attn:
            sa  = self.l2.self_attn
            xn  = self.l2.norm1(x)
            out, attn = sa(xn, xn, xn,
                           need_weights=True, average_attn_weights=True)
            x2  = x + self.l2.dropout1(out)
            xn2 = self.l2.norm2(x2)
            ff  = self.l2.linear2(self.l2.dropout(
                      self.l2.activation(self.l2.linear1(xn2))))
            return self.n2(x2 + self.l2.dropout2(ff))[:,0,:], attn
        return self.n2(self.l2(x))[:,0,:], None

# ── full model ────────────────────────────────────────────────────────────────

class RehabNet(nn.Module):
    """
    ST-GCN → Transformer-BiLSTM fusion
    Heads:
      classify_head : binary correct/incorrect
      quality_head  : continuous quality score 0-1
    Conditioning:
      mode_emb     : hardware mode (0-4)
      exercise_emb : exercise type — helps quality head learn
                     exercise-specific patterns
    """
    def __init__(self):
        super().__init__()
        self.stgcn        = STGCN()
        self.trans        = ClinicalTransformer()
        self.mode_emb     = nn.Embedding(N_MODES,      16)
        self.exercise_emb = nn.Embedding(N_EXERCISES,   8)  # NEW: fix 4
        self.shared       = nn.Sequential(
            nn.Linear(128 + 10 + 16 + 8, 128),  # +8 for exercise_emb
            nn.LayerNorm(128),
            nn.ReLU(),
            nn.Dropout(0.2))
        self.classify_head = nn.Linear(128, 2)
        self.quality_head  = nn.Linear(128, 1)

        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def encode(self, kps, adj, return_attn=False):
        if adj.dim() == 3: adj = adj[0]
        return self.trans(self.stgcn(kps, adj), return_attn)

    def forward(self, kps, adj, scalars, mode=None, exercise_idx=None):
        B = kps.shape[0]
        if mode is None:
            mode = torch.zeros(B, dtype=torch.long, device=kps.device)
        if exercise_idx is None:
            exercise_idx = torch.zeros(B, dtype=torch.long, device=kps.device)

        cls, _    = self.encode(kps, adj)
        mode_feat = self.mode_emb(mode)
        ex_feat   = self.exercise_emb(exercise_idx)
        x         = self.shared(torch.cat([cls, scalars, mode_feat, ex_feat], 1))
        logits    = self.classify_head(x)
        quality   = self.quality_head(x).squeeze(1)
        return logits, quality

# ── class-weighted loss helper ────────────────────────────────────────────────

def compute_class_weights(train_samples: List[dict],
                          device: torch.device) -> torch.Tensor:
    """
    Per-fold inverse-frequency class weights.
    Fixes s02-style collapse where one class dominates the training fold.
    """
    n_incorrect = sum(s['label'] == 0 for s in train_samples)
    n_correct   = sum(s['label'] == 1 for s in train_samples)
    total       = n_incorrect + n_correct
    w_incorrect = total / (2.0 * n_incorrect + 1e-6)
    w_correct   = total / (2.0 * n_correct   + 1e-6)
    return torch.tensor([w_incorrect, w_correct],
                        dtype=torch.float32, device=device)

# ── training ──────────────────────────────────────────────────────────────────

def train_epoch(model, loader, opt, device, class_weights, epoch=0):
    model.train()
    A       = torch.from_numpy(ADJ).to(device)
    total   = 0.0
    skipped = 0

    for batch_idx, b in enumerate(loader):
        kps   = b['keypoints'].to(device)
        lbl   = b['label'].to(device)
        qual  = b['quality'].to(device)
        sc    = b['scalars'].to(device)
        ex_idx= b['exercise_idx'].to(device)
        mode  = torch.zeros(kps.shape[0], dtype=torch.long, device=device)

        if torch.isnan(kps).any() or torch.isnan(sc).any():
            skipped += 1
            continue

        logits, qpred = model(kps, A, sc, mode, ex_idx)

        if torch.isnan(logits).any() or torch.isnan(qpred).any():
            if skipped < 3:
                print(f"  [ep{epoch}] NaN output batch {batch_idx} — skip")
            skipped += 1
            continue

        # fix 1: class-weighted cross entropy
        cls_loss  = F.cross_entropy(logits, lbl, weight=class_weights)
        qual_loss = F.mse_loss(torch.sigmoid(qpred), qual)
        loss      = cls_loss + 0.3 * qual_loss

        if torch.isnan(loss):
            skipped += 1
            continue

        opt.zero_grad()
        loss.backward()

        if epoch == 0 and batch_idx == 0:
            total_norm = sum(p.grad.norm()**2
                            for p in model.parameters()
                            if p.grad is not None) ** 0.5
            print(f"  [ep0 batch0] grad norm = {total_norm:.3f}")

        nn.utils.clip_grad_norm_(model.parameters(), 0.5)
        opt.step()
        total += loss.item()

    if skipped > 0:
        print(f"  epoch {epoch}: skipped {skipped} batches")
    return total / max(len(loader) - skipped, 1)

# ── evaluation ────────────────────────────────────────────────────────────────

@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    A = torch.from_numpy(ADJ).to(device)
    preds, labels, quals, qpreds = [], [], [], []

    for b in loader:
        kps    = b['keypoints'].to(device)
        sc     = b['scalars'].to(device)
        ex_idx = b['exercise_idx'].to(device)
        mode   = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        lg, qp = model(kps, A, sc, mode, ex_idx)
        preds.extend(lg.argmax(1).cpu().numpy())
        labels.extend(b['label'].numpy())
        quals.extend(b['quality'].numpy())
        qpreds.extend(torch.sigmoid(qp).cpu().numpy())

    from sklearn.metrics import accuracy_score, f1_score, classification_report
    from scipy.stats import pearsonr

    acc = accuracy_score(labels, preds)
    f1  = f1_score(labels, preds, average='macro', zero_division=0)
    r   = pearsonr(quals, qpreds)[0] if len(set(quals)) > 1 else 0.0
    print(classification_report(labels, preds,
          target_names=['incorrect','correct'], zero_division=0))
    print(f"  quality Pearson r={r:.3f}")
    return {'acc': acc, 'f1': f1, 'quality_r': r}

# ── LOSO training loop ────────────────────────────────────────────────────────

def run_loso(n_epochs=N_EPOCHS, batch_size=32, lr=3e-4, patience=12):
    """
    patience : early stopping — stop if val F1 doesn't improve for this
               many 5-epoch checkpoints (fix 2)
    """
    device   = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Device: {device}")
    samples  = build_master_dataset()
    subjects = get_uiprmd_subjects(samples)
    results  = []

    for subj in subjects:
        print(f"\n{'='*55}")
        print(f"LOSO held-out: {subj}")
        print(f"{'='*55}")
        tr, te = loso_split(samples, subj)
        tl, vl = build_loaders(tr, te, batch_size)

        # fix 1: compute per-fold class weights
        cw = compute_class_weights(tr, device)
        print(f"  class weights: incorrect={cw[0]:.3f}  correct={cw[1]:.3f}")

        model = RehabNet().to(device)
        opt   = torch.optim.AdamW(model.parameters(),
                                  lr=lr, weight_decay=1e-4)

        # fix 2: warmup + cosine
        def lr_lambda(ep):
            warmup = 5
            if ep < warmup:
                return (ep + 1) / warmup
            progress = (ep - warmup) / max(n_epochs - warmup, 1)
            return 0.5 * (1 + np.cos(np.pi * progress))

        sch = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda)

        # early stopping state
        best_f1    = 0.0
        best_state = None
        no_improve = 0

        for ep in range(n_epochs):
            loss = train_epoch(model, tl, opt, device, cw, epoch=ep)
            sch.step()

            # check every 5 epochs for early stopping
            if (ep + 1) % 5 == 0:
                # quick val F1
                model.eval()
                A   = torch.from_numpy(ADJ).to(device)
                p_list, l_list = [], []
                with torch.no_grad():
                    for b in vl:
                        kps    = b['keypoints'].to(device)
                        sc     = b['scalars'].to(device)
                        ex_idx = b['exercise_idx'].to(device)
                        mode   = torch.zeros(kps.shape[0],
                                             dtype=torch.long, device=device)
                        lg, _  = model(kps, A, sc, mode, ex_idx)
                        p_list.extend(lg.argmax(1).cpu().numpy())
                        l_list.extend(b['label'].numpy())
                model.train()

                from sklearn.metrics import f1_score as _f1
                val_f1 = _f1(l_list, p_list, average='macro', zero_division=0)
                print(f"  ep{ep+1:3d} | loss={loss:.4f} | "
                      f"lr={sch.get_last_lr()[0]:.2e} | val_f1={val_f1:.3f}")

                if val_f1 > best_f1:
                    best_f1    = val_f1
                    best_state = copy.deepcopy(model.state_dict())
                    no_improve = 0
                else:
                    no_improve += 1
                    if no_improve >= patience:
                        print(f"  Early stop at ep{ep+1} "
                              f"(no improvement for {patience*5} epochs)")
                        break

        # restore best checkpoint before final eval
        if best_state is not None:
            model.load_state_dict(best_state)
            print(f"  Restored best checkpoint (val_f1={best_f1:.3f})")

        print(f"\n── Evaluation: {subj} ──")
        results.append(evaluate(model, vl, device))

    accs = [r['acc'] for r in results]
    f1s  = [r['f1']  for r in results]
    qrs  = [r['quality_r'] for r in results]
    print(f"\n{'='*55}")
    print(f"LOSO FINAL RESULTS")
    print(f"  Accuracy  : {np.mean(accs):.3f} ± {np.std(accs):.3f}")
    print(f"  F1 macro  : {np.mean(f1s):.3f}  ± {np.std(f1s):.3f}")
    print(f"  Quality r : {np.mean(qrs):.3f}  ± {np.std(qrs):.3f}")
    print(f"{'='*55}")
    return results

# ── live inference ────────────────────────────────────────────────────────────

class LiveInference:
    def __init__(self, model: RehabNet, device='cpu',
                 active_mode=0, exercise='unknown'):
        self.model        = model.eval().to(device)
        self.device       = torch.device(device)
        self.buffer: List[np.ndarray] = []
        self.active_mode  = active_mode
        self.exercise_idx = EXERCISE2IDX.get(exercise, EXERCISE2IDX['unknown'])
        self.last_result  = None

    def set_mode(self, mode: int):
        assert mode in HARDWARE_MODES, f"mode must be 0-4, got {mode}"
        self.active_mode = mode
        print(f"Hardware mode → {HARDWARE_MODES[mode]['name']}")

    def set_exercise(self, exercise: str):
        self.exercise_idx = EXERCISE2IDX.get(exercise, EXERCISE2IDX['unknown'])
        print(f"Exercise → {exercise} (idx={self.exercise_idx})")

    def push_frame(self, ps1_frame: dict) -> Optional[dict]:
        if not validate_ps1_frame(ps1_frame):
            return None
        self.buffer.append(ps1_frame_to_keypoint(ps1_frame))
        if len(self.buffer) > TARGET_LEN:
            self.buffer = self.buffer[-TARGET_LEN:]
        if len(self.buffer) < TARGET_LEN:
            return None
        return self._infer()

    @torch.no_grad()
    def _infer(self) -> dict:
        raw  = np.stack(self.buffer, axis=0)
        kps  = process_keypoints(raw.copy())
        sc   = extract_scalars(kps)
        vel_proxy = float(sc[5])

        kps_t  = torch.from_numpy(kps).permute(2,0,1).unsqueeze(0).to(self.device)
        sc_t   = torch.from_numpy(sc).unsqueeze(0).to(self.device)
        A      = torch.from_numpy(ADJ).to(self.device)
        mode_t = torch.tensor([self.active_mode],
                               dtype=torch.long, device=self.device)
        ex_t   = torch.tensor([self.exercise_idx],
                               dtype=torch.long, device=self.device)

        logits, quality = self.model(kps_t, A, sc_t, mode_t, ex_t)
        prob    = F.softmax(logits, dim=1)[0]
        label   = int(logits.argmax(1).item())
        q       = float(torch.sigmoid(quality).item())
        torque  = compute_torque(q, self.active_mode, vel_proxy)

        self.last_result = {
            'label':         label,
            'correct_prob':  float(prob[1]),
            'quality_score': q,
            'torque_signal': torque,
            'mode':          self.active_mode,
            'mode_name':     HARDWARE_MODES[self.active_mode]['name'],
            'error_flag':    label == 0,
        }
        return self.last_result

# ── smoke test ────────────────────────────────────────────────────────────────

if __name__ == '__main__':
    print("── smoke test ──────────────────────────────────────────")
    B   = 4
    kps = torch.randn(B, 3, TARGET_LEN, N_JOINTS)
    adj = torch.from_numpy(ADJ).unsqueeze(0).expand(B,-1,-1)
    sc  = torch.rand(B, 10)
    ex  = torch.randint(0, N_EXERCISES, (B,))

    m = RehabNet()
    for mode_id in range(N_MODES):
        mode = torch.full((B,), mode_id, dtype=torch.long)
        lg, q = m(kps, adj, sc, mode, ex)
        print(f"  mode={mode_id} ({HARDWARE_MODES[mode_id]['name']:22s})"
              f"  logits={lg.shape}  quality={q.shape}")

    cls_vec, attn = m.encode(kps, adj, return_attn=True)
    print(f"  CLS={cls_vec.shape}  attn={attn.shape}")

    live = LiveInference(m, active_mode=2, exercise='deep_squat')
    for _ in range(TARGET_LEN + 5):
        frame = {f'{j}_{ax}': float(np.random.randn() * 0.3)
                 for j in ['HipLeft','KneeLeft','AnkleLeft',
                            'HipRight','KneeRight','AnkleRight']
                 for ax in ['x','y','z']}
        out = live.push_frame(frame)
    print(f"  live result: {out}")
    live.set_mode(4)
    live.set_exercise('squat')
    print("── passed ──────────────────────────────────────────────")
```

```python
import os
import copy
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d
from typing import Dict, List, Optional, Tuple
from collections import Counter

# â”€â”€ config â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

TARGET_LEN  = 150
N_JOINTS    = 6
IN_CHANNELS = 3
MIN_FRAMES  = 20
FPS         = 30
N_EPOCHS    = 60   # increased from 30

UIPRMD_CSV = '/content/drive/MyDrive/datasets/UIPRMD_timeseries.csv'
KIMORE_CSV = '/content/drive/MyDrive/datasets/squats_tabular_timeseries_binary.csv'
KERAAL_CSV = '/content/drive/MyDrive/datasets/keraal_final_binary.csv'

JOINT_ORDER = ['l_hip','l_knee','l_ankle','r_hip','r_knee','r_ankle']

EXERCISE_LIST = [
    'deep_squat','hurdle_step','inline_lunge','side_lunge',
    'sit_to_stand','straight_leg_raise','squat','ctk_squat','unknown'
]
EXERCISE2IDX = {e: i for i, e in enumerate(EXERCISE_LIST)}
N_EXERCISES  = len(EXERCISE_LIST)

# â”€â”€ hardware modes â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

HARDWARE_MODES = {
    0: {'name': 'free',              'scale': 0.0,  'bias': 0.0,  'min': 0.0, 'max': 0.0},
    1: {'name': 'constant_torque',   'scale': 0.0,  'bias': 0.5,  'min': 0.4, 'max': 0.6},
    2: {'name': 'resistive_torque',  'scale': 0.5,  'bias': 0.0,  'min': 0.0, 'max': 0.8},
    3: {'name': 'increasing_torque', 'scale': 1.0,  'bias': 0.0,  'min': 0.0, 'max': 1.0},
    4: {'name': 'dampening',         'scale': -0.5, 'bias': 0.5,  'min': 0.0, 'max': 0.5},
}
N_MODES = len(HARDWARE_MODES)

def compute_torque(quality: float, mode: int,
                   velocity_proxy: float = 0.0) -> float:
    m = HARDWARE_MODES.get(mode, HARDWARE_MODES[0])
    if mode == 4:
        torque = m['bias'] - abs(velocity_proxy) * 0.5
    else:
        torque = m['scale'] * quality + m['bias']
    return float(np.clip(torque, m['min'], m['max']))

# â”€â”€ adjacency â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def build_adjacency() -> np.ndarray:
    A = np.zeros((N_JOINTS, N_JOINTS), dtype=np.float32)
    for i, j in [(0,1),(1,2),(3,4),(4,5),(0,3),(1,4),(2,5)]:
        A[i,j] = A[j,i] = 1.0
    np.fill_diagonal(A, 1.0)
    d = np.diag(1.0 / np.sqrt(A.sum(axis=1) + 1e-6))
    return (d @ A @ d).astype(np.float32)

ADJ = build_adjacency()

EPS = 1e-6

def safe_np(x, fill=0.0, clip=None, dtype=np.float32):
    x = np.asarray(x, dtype=dtype)
    x = np.nan_to_num(x, nan=fill, posinf=fill, neginf=fill)
    if clip is not None:
        x = np.clip(x, -clip, clip)
    return x.astype(dtype)

def valid_np(x) -> bool:
    x = np.asarray(x)
    return x.size > 0 and np.isfinite(x).all()

# â”€â”€ signal helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def _smooth(arr, win=7, poly=2):
    arr = safe_np(arr)
    if len(arr) < win + 2:
        return arr
    w = win if win % 2 == 1 else win + 1
    w = max(w, poly+2 if (poly+2)%2==1 else poly+3)
    try:
        return safe_np(savgol_filter(arr, w, poly))
    except Exception:
        return arr

def _resample(arr, n):
    arr = safe_np(arr)
    if len(arr) == n:
        return arr
    if len(arr) < 2:
        return np.zeros(n, dtype=np.float32)
    return safe_np(np.interp(np.linspace(0,1,n), np.linspace(0,1,len(arr)), arr))

def _fill_missing(kps):
    kps = safe_np(kps)
    T, J, C = kps.shape
    for j in range(J):
        for c in range(C):
            col = kps[:,j,c]
            z   = (~np.isfinite(col)) | (col == 0.0)
            if z.all():
                opp = j+3 if j < 3 else j-3
                kps[:,j,c] = kps[:,opp,c]
            elif z.any() and (~z).sum() >= 2:
                f = interp1d(np.where(~z)[0], col[~z],
                             bounds_error=False, fill_value='extrapolate')
                kps[:,j,c] = f(np.arange(T))
    return safe_np(kps)

def _normalise(kps):
    kps = safe_np(kps)
    hip   = ((kps[:,0,:] + kps[:,3,:]) / 2.0).mean(axis=0)
    kps   = kps - hip
    l     = np.mean(np.linalg.norm(kps[:,0,:] - kps[:,2,:], axis=1))
    r     = np.mean(np.linalg.norm(kps[:,3,:] - kps[:,5,:], axis=1))
    scale = (l + r) / 2.0
    if (not np.isfinite(scale)) or scale < 1e-4:
        scale = 1.0
    kps   = kps / scale
    kps   = np.clip(kps, -5.0, 5.0)
    return safe_np(kps, clip=5.0)

def process_keypoints(raw: np.ndarray) -> np.ndarray:
    raw = safe_np(raw)
    if raw.ndim != 3 or raw.shape[1] != N_JOINTS or raw.shape[2] != 3:
        raise ValueError(f"Expected raw shape (T,{N_JOINTS},3), got {raw.shape}")
    if len(raw) < MIN_FRAMES:
        raise ValueError(f"Too short: {len(raw)} frames")
    for j in range(N_JOINTS):
        for c in range(3):
            raw[:,j,c] = _smooth(raw[:,j,c])
    raw = _fill_missing(raw)
    out = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
    for j in range(N_JOINTS):
        for c in range(3):
            out[:,j,c] = _resample(raw[:,j,c], TARGET_LEN)
    out = _normalise(out)
    return safe_np(out, clip=5.0)

# â”€â”€ scalar features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def extract_scalars(kps: np.ndarray) -> np.ndarray:
    kps = safe_np(kps, clip=5.0)

    def _ang(a, b, c):
        ba = a - b;  bc = c - b
        denom = np.maximum(
            np.linalg.norm(ba, axis=1) * np.linalg.norm(bc, axis=1),
            EPS)
        cos = np.sum(ba*bc, axis=1) / denom
        return np.degrees(np.arccos(np.clip(cos, -1, 1)))

    lk = _ang(kps[:,0,:], kps[:,1,:], kps[:,2,:])
    rk = _ang(kps[:,3,:], kps[:,4,:], kps[:,5,:])

    lr  = float(np.nanmax(lk) - np.nanmin(lk))
    rr  = float(np.nanmax(rk) - np.nanmin(rk))
    sym = abs(lr-rr) / ((lr+rr)/2.0+1e-6) * 100.0 if (lr+rr) > 1.0 else 0.0

    vel_arr = np.abs(np.diff(lk)) * FPS
    vel     = float(np.max(vel_arr)) if len(vel_arr) > 0 else 0.0

    hip  = (kps[:,0,:] + kps[:,3,:]) / 2.0
    jerk_arr = np.diff(np.diff(hip, axis=0), axis=0)
    jerk = float(np.nanmean(np.linalg.norm(jerk_arr, axis=1))) if len(jerk_arr) else 0.0

    peak  = int(np.argmin(lk))
    trunk = float(abs(hip[peak, 0] - hip[0, 0]))

    lk_c = lk - np.nanmean(lk);  rk_c = rk - np.nanmean(rk)
    if np.nanstd(lk_c) > 1e-6 and np.nanstd(rk_c) > 1e-6:
        lag = float(abs(
            np.argmax(np.correlate(lk_c, rk_c, 'full')) - (len(lk)-1)))
    else:
        lag = 0.0

    diff_std = float(np.nanstd(np.diff(lk)))
    smooth   = float(np.clip(1.0 / (diff_std + 1e-6) / 100.0, 0.0, 1.0))

    scalars = np.array([
        np.clip(lr   / 180.0, 0, 1),
        np.clip(rr   / 180.0, 0, 1),
        np.clip(np.min(lk) / 180.0, 0, 1),
        np.clip(np.min(rk) / 180.0, 0, 1),
        np.clip(sym  / 100.0, 0, 5),
        np.clip(vel  / 200.0, 0, 5),
        np.clip(jerk / 10.0,  0, 5),
        np.clip(trunk,        0, 5),
        np.clip(lag  / TARGET_LEN, 0, 1),
        smooth,
    ], dtype=np.float32)

    return safe_np(scalars, fill=0.0, clip=10.0)

# â”€â”€ PS1 interface â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def validate_ps1_frame(frame: dict) -> bool:
    for j in ['HipLeft','KneeLeft','AnkleLeft',
              'HipRight','KneeRight','AnkleRight']:
        for ax in ['x','y','z']:
            v = frame.get(f'{j}_{ax}')
            if v is None or np.isnan(float(v)) or abs(float(v)) > 10.0:
                return False
    return True

def ps1_frame_to_keypoint(frame: dict) -> np.ndarray:
    row = np.zeros((N_JOINTS, 3), dtype=np.float32)
    for ji, jname in enumerate(['HipLeft','KneeLeft','AnkleLeft',
                                  'HipRight','KneeRight','AnkleRight']):
        for ci, ax in enumerate(['x','y','z']):
            row[ji, ci] = float(frame.get(f'{jname}_{ax}', 0.0))
    return row

# â”€â”€ dataset loaders â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def _kps_from_df_rows(df, xyz_cols):
    return df[xyz_cols].values.astype(np.float32).reshape(len(df), N_JOINTS, 3)

def load_uiprmd() -> List[dict]:
    if not os.path.exists(UIPRMD_CSV):
        print(f"UI-PRMD CSV not found: {UIPRMD_CSV}"); return []
    df       = pd.read_csv(UIPRMD_CSV)
    xyz_cols = [f'{j}_{ax}' for j in JOINT_ORDER for ax in ['x','y','z']]
    samples  = []
    for (subj, mov, ex_id, label_val), grp in df.groupby(
            ['subject','movement','exercise_id','label']):
        grp   = grp.sort_values('frame')
        label = int(label_val)
        raw   = _kps_from_df_rows(grp, xyz_cols) / 1000.0
        try:
            kps = process_keypoints(raw)
            ex  = grp['exercise'].iloc[0]
            samples.append({
                'keypoints':   kps,
                'adj':         ADJ,
                'label':       label,
                'quality':     float(label),
                'exercise':    ex,
                'exercise_idx': EXERCISE2IDX.get(ex, EXERCISE2IDX['unknown']),
                'subject':     f'uiprmd_{subj}',
                'source':      'uiprmd',
                'scalars':     extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip uiprmd {subj}/{mov}/{ex_id}: {e}")
    print(f"UI-PRMD: {len(samples)} samples")
    return samples

def load_kimore() -> List[dict]:
    if not os.path.exists(KIMORE_CSV):
        print(f"KIMORE not found: {KIMORE_CSV}"); return []
    df = pd.read_csv(KIMORE_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    df['quality_norm'] = df[['cTS','cPO','cCF']].mean(axis=1) / 50.0
    samples = []
    for subj, grp in df.groupby('Subject'):
        grp     = grp.sort_values('FrameID')
        label   = int(grp['label'].iloc[0])
        quality = float(grp['quality_norm'].mean())
        raw     = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps = process_keypoints(raw)
            samples.append({
                'keypoints':    kps,
                'adj':          ADJ,
                'label':        label,
                'quality':      quality,
                'exercise':     'squat',
                'exercise_idx': EXERCISE2IDX['squat'],
                'subject':      f'kimore_{subj}',
                'source':       'kimore',
                'scalars':      extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip kimore {subj}: {e}")
    print(f"KIMORE: {len(samples)} samples")
    return samples

def load_keraal() -> List[dict]:
    if not os.path.exists(KERAAL_CSV):
        print(f"Keraal not found: {KERAAL_CSV}"); return []
    df = pd.read_csv(KERAAL_CSV)
    xyz_cols = []
    for jname in ['HipLeft','KneeLeft','AnkleLeft',
                  'HipRight','KneeRight','AnkleRight']:
        xyz_cols += [f'{jname}_x', f'{jname}_y', f'{jname}_z']
    samples = []
    for ann_key, grp in df.groupby('ann_key'):
        grp   = grp.sort_values('FrameID') if 'FrameID' in grp.columns else grp
        label = int(grp['label'].iloc[0])
        raw   = _kps_from_df_rows(grp, xyz_cols)
        try:
            kps  = process_keypoints(raw)
            subj = '-'.join(str(ann_key).split('-')[:3])
            samples.append({
                'keypoints':    kps,
                'adj':          ADJ,
                'label':        label,
                'quality':      float(label),
                'exercise':     'ctk_squat',
                'exercise_idx': EXERCISE2IDX['ctk_squat'],
                'subject':      f'keraal_{subj}',
                'source':       'keraal',
                'scalars':      extract_scalars(kps),
            })
        except Exception as e:
            print(f"  skip keraal {ann_key}: {e}")
    print(f"Keraal: {len(samples)} samples")
    return samples

def build_master_dataset() -> List[dict]:
    samples = load_uiprmd() + load_kimore() + load_keraal()
    clean   = []
    dropped_by_source = {}
    for s in samples:
        source = s.get('source', 'unknown')
        try:
            kps = safe_np(s['keypoints'], clip=5.0)
            scalars = safe_np(s.get('scalars', extract_scalars(kps)), clip=10.0)
            quality = float(np.nan_to_num(s.get('quality', s.get('label', 1)),
                                          nan=float(s.get('label', 1))))

            if kps.shape != (TARGET_LEN, N_JOINTS, 3):
                raise ValueError(f"bad keypoint shape {kps.shape}")
            if scalars.shape != (10,):
                raise ValueError(f"bad scalar shape {scalars.shape}")
            if not valid_np(kps) or not valid_np(scalars):
                raise ValueError("NaN/Inf after sanitizing")

            s['keypoints'] = kps
            s['scalars'] = scalars
            s['label'] = int(s.get('label', 1))
            s['quality'] = float(np.clip(quality, 0.0, 1.0))
            clean.append(s)
        except Exception as e:
            print(f"  DROP corrupt: {s.get('subject', 'unknown')} "
                  f"{s.get('exercise', 'unknown')} ({e})")
            dropped_by_source[source] = dropped_by_source.get(source, 0) + 1
            continue
    print(f"\nMaster: {len(clean)} samples "
          f"(dropped {len(samples)-len(clean)} corrupt)")
    print(f"  correct={sum(s['label']==1 for s in clean)}"
          f"  incorrect={sum(s['label']==0 for s in clean)}")
    print(f"  sources:   {set(s['source']   for s in clean)}")
    print(f"  exercises: {set(s['exercise'] for s in clean)}")
    if dropped_by_source:
        print(f"  dropped_by_source: {dropped_by_source}")
    return clean

# â”€â”€ LOSO helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def loso_split(samples, held_out):
    return ([s for s in samples if s['subject'] != held_out],
            [s for s in samples if s['subject'] == held_out])

def get_uiprmd_subjects(samples):
    return sorted({s['subject'] for s in samples if s['source'] == 'uiprmd'})

# â”€â”€ PyTorch dataset â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class RehabDataset(Dataset):
    def __init__(self, samples, augment=False):
        self.samples = samples
        self.augment = augment

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        s   = self.samples[idx]
        kps = safe_np(s['keypoints'], clip=5.0)
        if self.augment:
            kps = self._aug(kps)
        kps = safe_np(kps, clip=5.0)
        scalars = safe_np(s['scalars'], clip=10.0)
        return {
            'keypoints':    torch.from_numpy(kps).permute(2,0,1).float(),
            'adj':          torch.from_numpy(ADJ).float(),
            'label':        torch.tensor(s['label'],        dtype=torch.long),
            'quality':      torch.tensor(s['quality'],      dtype=torch.float32),
            'scalars':      torch.from_numpy(scalars).float(),
            'exercise_idx': torch.tensor(s['exercise_idx'], dtype=torch.long),
            'exercise':     s['exercise'],
            'subject':      s['subject'],
        }

    def _aug(self, kps):
        kps = safe_np(kps, clip=5.0)
        if np.random.rand() < 0.5:
            n = max(int(TARGET_LEN * np.random.uniform(0.8, 1.2)), MIN_FRAMES)
            tmp = np.zeros((n, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    tmp[:,j,c] = _resample(kps[:,j,c], n)
            kps = np.zeros((TARGET_LEN, N_JOINTS, 3), dtype=np.float32)
            for j in range(N_JOINTS):
                for c in range(3):
                    kps[:,j,c] = _resample(tmp[:,j,c], TARGET_LEN)
        if np.random.rand() < 0.5:
            kps[:,[0,1,2,3,4,5],:] = kps[:,[3,4,5,0,1,2],:]
            kps[:,:,0] *= -1.0
        if np.random.rand() < 0.5:
            kps += np.random.randn(*kps.shape).astype(np.float32) * 0.02
        if np.random.rand() < 0.3:
            th  = np.radians(np.random.uniform(-10, 10))
            x   = kps[:,:,0].copy()
            z   = kps[:,:,2].copy()
            kps[:,:,0] = np.cos(th)*x - np.sin(th)*z
            kps[:,:,2] = np.sin(th)*x + np.cos(th)*z
        return safe_np(kps, clip=5.0)

def collate(batch):
    return {k: torch.stack([b[k] for b in batch])
              if isinstance(batch[0][k], torch.Tensor)
              else [b[k] for b in batch]
            for k in batch[0]}

def build_loaders(train_samples, test_samples, batch_size=32):
    from torch.utils.data import WeightedRandomSampler
    ex_counts = Counter(s['exercise'] for s in train_samples)
    total     = len(train_samples)
    weights   = [total / ex_counts[s['exercise']] for s in train_samples]
    sampler   = WeightedRandomSampler(weights, len(train_samples), replacement=True)
    return (
        DataLoader(RehabDataset(train_samples, True),  batch_size,
                   sampler=sampler, collate_fn=collate, num_workers=2),
        DataLoader(RehabDataset(test_samples,  False), batch_size, False,
                   collate_fn=collate, num_workers=2),
    )

# â”€â”€ ST-GCN â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class STGCNBlock(nn.Module):
    def __init__(self, c_in, c_out, ks=9, stride=1):
        super().__init__()
        self.gcn = nn.Linear(c_in, c_out)
        self.tcn = nn.Sequential(
            nn.BatchNorm2d(c_out), nn.ReLU(),
            nn.Conv2d(c_out, c_out, (ks,1), (stride,1), ((ks-1)//2, 0)),
            nn.BatchNorm2d(c_out), nn.Dropout(0.1))
        self.res = (nn.Sequential(nn.Conv2d(c_in, c_out, 1, (stride,1)),
                                   nn.BatchNorm2d(c_out))
                    if c_in != c_out or stride != 1 else nn.Identity())
        self.relu = nn.ReLU()

    def forward(self, x, A):
        B, C, T, V = x.shape
        if A.dim() == 3: A = A[0]
        xs      = x.permute(0,2,3,1).contiguous().view(B*T, V, C)
        xs      = torch.bmm(A.unsqueeze(0).expand(B*T,-1,-1), xs)
        xs      = self.gcn(xs).view(B, T, V, -1).permute(0,3,1,2)
        tcn_out = self.tcn(xs)
        res_out = self.res(x)
        if torch.isnan(tcn_out).any() or torch.isnan(res_out).any():
            tcn_out = torch.nan_to_num(tcn_out, 0.0)
            res_out = torch.nan_to_num(res_out, 0.0)
        return self.relu(tcn_out + res_out)

class STGCN(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_norm = nn.LayerNorm(IN_CHANNELS * N_JOINTS)
        self.b1 = STGCNBlock(3,  32)
        self.b2 = STGCNBlock(32, 64)
        self.b3 = STGCNBlock(64, 128)

    def forward(self, x, A):
        B, C, T, V = x.shape
        x  = torch.clamp(x, -5.0, 5.0)
        xf = x.permute(0,2,1,3).contiguous().view(B*T, C*V)
        xf = self.input_norm(xf).view(B, T, C, V).permute(0,2,1,3)
        return self.b3(self.b2(self.b1(xf, A), A), A).mean(3).permute(0,2,1)

# â”€â”€ clinical transformer â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class ClinicalTransformer(nn.Module):
    def __init__(self, d=128, heads=4, d_ff=256, drop=0.1):
        super().__init__()
        pe  = torch.zeros(TARGET_LEN+1, d)
        pos = torch.arange(TARGET_LEN+1).unsqueeze(1).float()
        div = torch.exp(torch.arange(0,d,2).float() * (-np.log(10000.0)/d))
        pe[:,0::2] = torch.sin(pos*div)
        pe[:,1::2] = torch.cos(pos*div)
        self.register_buffer('pe', pe.unsqueeze(0))
        self.cls  = nn.Parameter(torch.randn(1,1,d) * 0.02)
        self.drop = nn.Dropout(drop)
        mk = lambda: nn.TransformerEncoderLayer(
            d, heads, d_ff, drop, batch_first=True, norm_first=True)
        self.l1, self.l2 = mk(), mk()
        self.n1, self.n2 = nn.LayerNorm(d), nn.LayerNorm(d)

    def forward(self, x, return_attn=False):
        B, T, _ = x.shape
        x = torch.cat([self.cls.expand(B,-1,-1), x], 1)
        x = self.drop(x + self.pe[:, :T+1, :])
        x = self.n1(self.l1(x))
        if return_attn:
            sa  = self.l2.self_attn
            xn  = self.l2.norm1(x)
            out, attn = sa(xn, xn, xn,
                           need_weights=True, average_attn_weights=True)
            x2  = x + self.l2.dropout1(out)
            xn2 = self.l2.norm2(x2)
            ff  = self.l2.linear2(self.l2.dropout(
                      self.l2.activation(self.l2.linear1(xn2))))
            return self.n2(x2 + self.l2.dropout2(ff))[:,0,:], attn
        return self.n2(self.l2(x))[:,0,:], None

# â”€â”€ full model â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class RehabNet(nn.Module):
    """
    ST-GCN â†’ Transformer-BiLSTM fusion
    Heads:
      classify_head : binary correct/incorrect
      quality_head  : continuous quality score 0-1
    Conditioning:
      mode_emb     : hardware mode (0-4)
      exercise_emb : exercise type â€” helps quality head learn
                     exercise-specific patterns
    """
    def __init__(self):
        super().__init__()
        self.stgcn        = STGCN()
        self.trans        = ClinicalTransformer()
        self.mode_emb     = nn.Embedding(N_MODES,      16)
        self.exercise_emb = nn.Embedding(N_EXERCISES,   8)  # NEW: fix 4
        self.shared       = nn.Sequential(
            nn.Linear(128 + 10 + 16 + 8, 128),  # +8 for exercise_emb
            nn.LayerNorm(128),
            nn.ReLU(),
            nn.Dropout(0.2))
        self.classify_head = nn.Linear(128, 2)
        self.quality_head  = nn.Linear(128, 1)

        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def encode(self, kps, adj, return_attn=False):
        if adj.dim() == 3: adj = adj[0]
        return self.trans(self.stgcn(kps, adj), return_attn)

    def forward(self, kps, adj, scalars, mode=None, exercise_idx=None):
        B = kps.shape[0]
        if mode is None:
            mode = torch.zeros(B, dtype=torch.long, device=kps.device)
        if exercise_idx is None:
            exercise_idx = torch.zeros(B, dtype=torch.long, device=kps.device)

        cls, _    = self.encode(kps, adj)
        mode_feat = self.mode_emb(mode)
        ex_feat   = self.exercise_emb(exercise_idx)
        x         = self.shared(torch.cat([cls, scalars, mode_feat, ex_feat], 1))
        logits    = self.classify_head(x)
        quality   = self.quality_head(x).squeeze(1)
        return logits, quality

#  class-weighted loss helper

def compute_class_weights(train_samples: List[dict],
                          device: torch.device) -> torch.Tensor:
    """
    Per-fold inverse-frequency class weights.
    Fixes s02-style collapse where one class dominates the training fold.
    """
    n_incorrect = sum(s['label'] == 0 for s in train_samples)
    n_correct   = sum(s['label'] == 1 for s in train_samples)
    total       = n_incorrect + n_correct
    w_incorrect = total / (2.0 * n_incorrect + 1e-6)
    w_correct   = total / (2.0 * n_correct   + 1e-6)
    return torch.tensor([w_incorrect, w_correct],
                        dtype=torch.float32, device=device)

# training

def train_epoch(model, loader, opt, device, class_weights, epoch=0):
    model.train()
    A = torch.from_numpy(ADJ).to(device)
    total = 0.0
    used_batches = 0
    skipped_batches = 0
    dropped_samples = 0

    for batch_idx, b in enumerate(loader):
        kps = b['keypoints'].to(device)
        lbl = b['label'].to(device)
        qual = b['quality'].to(device)
        sc = b['scalars'].to(device)
        ex_idx = b['exercise_idx'].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1) &
            torch.isfinite(sc).all(1) &
            torch.isfinite(qual) &
            torch.isfinite(ex_idx.float())
        )

        if not good.all():
            dropped_samples += int((~good).sum().item())
            kps = kps[good]
            lbl = lbl[good]
            qual = qual[good]
            sc = sc[good]
            ex_idx = ex_idx[good]

        if kps.shape[0] == 0:
            skipped_batches += 1
            continue

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        logits, qpred = model(kps, A, sc, mode, ex_idx)

        if not torch.isfinite(logits).all() or not torch.isfinite(qpred).all():
            if skipped_batches < 3:
                print(f"  [ep{epoch}] unstable output batch {batch_idx} - skip")
            skipped_batches += 1
            continue

        cls_loss = F.cross_entropy(logits, lbl, weight=class_weights)
        qual = torch.clamp(qual, 0.0, 1.0)
        qual_loss = F.mse_loss(torch.sigmoid(qpred), qual)
        loss = cls_loss + 0.3 * qual_loss

        if not torch.isfinite(loss):
            skipped_batches += 1
            continue

        opt.zero_grad(set_to_none=True)
        loss.backward()

        if epoch == 0 and batch_idx == 0:
            total_norm = sum(p.grad.norm()**2
                            for p in model.parameters()
                            if p.grad is not None) ** 0.5
            print(f"  [ep0 batch0] grad norm = {total_norm:.3f}")

        nn.utils.clip_grad_norm_(model.parameters(), 0.5)
        opt.step()
        total += loss.item()
        used_batches += 1

    if dropped_samples > 0 or skipped_batches > 0:
        print(f"  epoch {epoch}: dropped {dropped_samples} bad samples, "
              f"skipped {skipped_batches} empty/unstable batches")
    return total / max(used_batches, 1)

@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    A = torch.from_numpy(ADJ).to(device)
    preds, labels, quals, qpreds = [], [], [], []

    for b in loader:
        kps = b['keypoints'].to(device)
        sc = b['scalars'].to(device)
        ex_idx = b['exercise_idx'].to(device)
        qual = b['quality'].to(device)
        lbl = b['label'].to(device)

        good = (
            torch.isfinite(kps).flatten(1).all(1) &
            torch.isfinite(sc).all(1) &
            torch.isfinite(qual) &
            torch.isfinite(ex_idx.float())
        )
        if good.sum().item() == 0:
            continue

        kps = kps[good]
        sc = sc[good]
        ex_idx = ex_idx[good]
        qual = qual[good]
        lbl = lbl[good]

        mode = torch.zeros(kps.shape[0], dtype=torch.long, device=device)
        lg, qp = model(kps, A, sc, mode, ex_idx)
        preds.extend(lg.argmax(1).cpu().numpy())
        labels.extend(lbl.cpu().numpy())
        quals.extend(qual.cpu().numpy())
        qpreds.extend(torch.sigmoid(qp).cpu().numpy())

    from sklearn.metrics import accuracy_score, f1_score, classification_report
    from scipy.stats import pearsonr

    if len(labels) == 0:
        print("  WARNING: no valid evaluation samples")
        return {'acc': 0.0, 'f1': 0.0, 'quality_r': 0.0}

    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='macro', zero_division=0)
    r = pearsonr(quals, qpreds)[0] if len(set(quals)) > 1 else 0.0
    print(classification_report(labels, preds,
          target_names=['incorrect','correct'], zero_division=0))
    print(f"  quality Pearson r={r:.3f}")
    return {'acc': acc, 'f1': f1, 'quality_r': r}
# LOSO training loop

def run_loso(n_epochs=N_EPOCHS, batch_size=32, lr=3e-4, patience=12):
    """
    patience : early stopping â€” stop if val F1 doesn't improve for this
               many 5-epoch checkpoints (fix 2)
    """
    device   = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Device: {device}")
    samples  = build_master_dataset()
    subjects = get_uiprmd_subjects(samples)
    results  = []

    for subj in subjects:
        print(f"\n{'='*55}")
        print(f"LOSO held-out: {subj}")
        print(f"{'='*55}")
        tr, te = loso_split(samples, subj)
        tl, vl = build_loaders(tr, te, batch_size)

        # fix 1: compute per-fold class weights
        cw = compute_class_weights(tr, device)
        print(f"  class weights: incorrect={cw[0]:.3f}  correct={cw[1]:.3f}")

        model = RehabNet().to(device)
        opt   = torch.optim.AdamW(model.parameters(),
                                  lr=lr, weight_decay=1e-4)

        # fix 2: warmup + cosine
        def lr_lambda(ep):
            warmup = 5
            if ep < warmup:
                return (ep + 1) / warmup
            progress = (ep - warmup) / max(n_epochs - warmup, 1)
            return 0.5 * (1 + np.cos(np.pi * progress))

        sch = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda)

        # early stopping state
        best_f1    = 0.0
        best_state = None
        no_improve = 0

        for ep in range(n_epochs):
            loss = train_epoch(model, tl, opt, device, cw, epoch=ep)
            sch.step()

            # check every 5 epochs for early stopping
            if (ep + 1) % 5 == 0:
                # quick val F1
                model.eval()
                A   = torch.from_numpy(ADJ).to(device)
                p_list, l_list = [], []
                with torch.no_grad():
                    for b in vl:
                        kps = b['keypoints'].to(device)
                        sc = b['scalars'].to(device)
                        ex_idx = b['exercise_idx'].to(device)
                        lbl = b['label'].to(device)
                        good = (
                            torch.isfinite(kps).flatten(1).all(1) &
                            torch.isfinite(sc).all(1) &
                            torch.isfinite(ex_idx.float())
                        )
                        if good.sum().item() == 0:
                            continue
                        kps = kps[good]
                        sc = sc[good]
                        ex_idx = ex_idx[good]
                        lbl = lbl[good]
                        mode = torch.zeros(kps.shape[0],
                                           dtype=torch.long, device=device)
                        lg, _ = model(kps, A, sc, mode, ex_idx)
                        p_list.extend(lg.argmax(1).cpu().numpy())
                        l_list.extend(lbl.cpu().numpy())
                model.train()

                from sklearn.metrics import f1_score as _f1
                val_f1 = _f1(l_list, p_list, average='macro', zero_division=0)
                print(f"  ep{ep+1:3d} | loss={loss:.4f} | "
                      f"lr={sch.get_last_lr()[0]:.2e} | val_f1={val_f1:.3f}")

                if val_f1 > best_f1:
                    best_f1    = val_f1
                    best_state = copy.deepcopy(model.state_dict())
                    no_improve = 0
                else:
                    no_improve += 1
                    if no_improve >= patience:
                        print(f"  Early stop at ep{ep+1} "
                              f"(no improvement for {patience*5} epochs)")
                        break

        # restore best checkpoint before final eval
        if best_state is not None:
            model.load_state_dict(best_state)
            print(f"  Restored best checkpoint (val_f1={best_f1:.3f})")

        print(f"\nâ”€â”€ Evaluation: {subj} â”€â”€")
        results.append(evaluate(model, vl, device))

    accs = [r['acc'] for r in results]
    f1s  = [r['f1']  for r in results]
    qrs  = [r['quality_r'] for r in results]
    print(f"\n{'='*55}")
    print(f"LOSO FINAL RESULTS")
    print(f"  Accuracy  : {np.mean(accs):.3f} Â± {np.std(accs):.3f}")
    print(f"  F1 macro  : {np.mean(f1s):.3f}  Â± {np.std(f1s):.3f}")
    print(f"  Quality r : {np.mean(qrs):.3f}  Â± {np.std(qrs):.3f}")
    print(f"{'='*55}")
    return results

# live inference

class LiveInference:
    def __init__(self, model: RehabNet, device='cpu',
                 active_mode=0, exercise='unknown'):
        self.model        = model.eval().to(device)
        self.device       = torch.device(device)
        self.buffer: List[np.ndarray] = []
        self.active_mode  = active_mode
        self.exercise_idx = EXERCISE2IDX.get(exercise, EXERCISE2IDX['unknown'])
        self.last_result  = None

    def set_mode(self, mode: int):
        assert mode in HARDWARE_MODES, f"mode must be 0-4, got {mode}"
        self.active_mode = mode
        print(f"Hardware mode â†’ {HARDWARE_MODES[mode]['name']}")

    def set_exercise(self, exercise: str):
        self.exercise_idx = EXERCISE2IDX.get(exercise, EXERCISE2IDX['unknown'])
        print(f"Exercise â†’ {exercise} (idx={self.exercise_idx})")

    def push_frame(self, ps1_frame: dict) -> Optional[dict]:
        if not validate_ps1_frame(ps1_frame):
            return None
        self.buffer.append(ps1_frame_to_keypoint(ps1_frame))
        if len(self.buffer) > TARGET_LEN:
            self.buffer = self.buffer[-TARGET_LEN:]
        if len(self.buffer) < TARGET_LEN:
            return None
        return self._infer()

    @torch.no_grad()
    def _infer(self) -> dict:
        raw  = np.stack(self.buffer, axis=0)
        kps  = process_keypoints(raw.copy())
        sc   = extract_scalars(kps)
        vel_proxy = float(sc[5])

        kps_t  = torch.from_numpy(kps).permute(2,0,1).unsqueeze(0).to(self.device)
        sc_t   = torch.from_numpy(sc).unsqueeze(0).to(self.device)
        A      = torch.from_numpy(ADJ).to(self.device)
        mode_t = torch.tensor([self.active_mode],
                               dtype=torch.long, device=self.device)
        ex_t   = torch.tensor([self.exercise_idx],
                               dtype=torch.long, device=self.device)

        logits, quality = self.model(kps_t, A, sc_t, mode_t, ex_t)
        prob    = F.softmax(logits, dim=1)[0]
        label   = int(logits.argmax(1).item())
        q       = float(torch.sigmoid(quality).item())
        torque  = compute_torque(q, self.active_mode, vel_proxy)

        self.last_result = {
            'label':         label,
            'correct_prob':  float(prob[1]),
            'quality_score': q,
            'torque_signal': torque,
            'mode':          self.active_mode,
            'mode_name':     HARDWARE_MODES[self.active_mode]['name'],
            'error_flag':    label == 0,
        }
        return self.last_result

# smoke test

if __name__ == '__main__':
    print("   smoke test ")
    B   = 4
    kps = torch.randn(B, 3, TARGET_LEN, N_JOINTS)
    adj = torch.from_numpy(ADJ).unsqueeze(0).expand(B,-1,-1)
    sc  = torch.rand(B, 10)
    ex  = torch.randint(0, N_EXERCISES, (B,))

    m = RehabNet()
    for mode_id in range(N_MODES):
        mode = torch.full((B,), mode_id, dtype=torch.long)
        lg, q = m(kps, adj, sc, mode, ex)
        print(f"  mode={mode_id} ({HARDWARE_MODES[mode_id]['name']:22s})"
              f"  logits={lg.shape}  quality={q.shape}")

    cls_vec, attn = m.encode(kps, adj, return_attn=True)
    print(f"  CLS={cls_vec.shape}  attn={attn.shape}")

    live = LiveInference(m, active_mode=2, exercise='deep_squat')
    for _ in range(TARGET_LEN + 5):
        frame = {f'{j}_{ax}': float(np.random.randn() * 0.3)
                 for j in ['HipLeft','KneeLeft','AnkleLeft',
                            'HipRight','KneeRight','AnkleRight']
                 for ax in ['x','y','z']}
        out = live.push_frame(frame)
    print(f"  live result: {out}")
    live.set_mode(4)
    live.set_exercise('squat')
    print("passed")

```

```python
"""
=============================================================================
SA-RNN — Bidirectional LSTM + Self-Attention for PS2 comparison
=============================================================================
Plugs directly into ps2_pipeline.py's data pipeline.
Runs same LOSO loop as RehabNet for fair comparison.

Three models compared in ablation:
  1. RehabNet       — ST-GCN + Two-Layer Clinical Transformer  (your main model)
  2. SARNN          — ST-GCN + BiLSTM + Self-Attention         (this file)
  3. LSTMOnly       — ST-GCN + BiLSTM, no attention            (ablation baseline)

Why BiLSTM + attention beats vanilla LSTM for rehab:
  - Bidirectional: descent phase informs interpretation of ascent and vice versa
  - Self-attention: learns which specific frames (e.g. peak valgus moment)
    drive the classification decision — matches your therapist dashboard heatmap
  - Lighter than Transformer: fewer parameters, faster to train, better on
    small datasets like UI-PRMD

Usage in Colab:
  from sarnn_model import run_ablation
  ablation_results = run_ablation(n_epochs=30, batch_size=32)
=============================================================================
"""

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from typing import List, Tuple, Optional, Dict
from sklearn.metrics import accuracy_score, f1_score, classification_report

# ─── Self-Attention module ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

class SelfAttention(nn.Module):
    def __init__(self, hidden_dim: int):
        super().__init__()
        self.attn  = nn.Linear(hidden_dim, hidden_dim)
        self.query = nn.Linear(hidden_dim, 1, bias=False)

    def forward(self, x: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
        energy  = torch.tanh(self.attn(x))
        scores  = self.query(energy).squeeze(-1)
        weights = F.softmax(scores, dim=1)
        context = torch.bmm(weights.unsqueeze(1), x).squeeze(1)
        return context, weights

# ─── SA-RNN model ━━━━━━━━━━━━━━━━━━━━━━━━

class SARNN(nn.Module):
    def __init__(self, n_classes=2, lstm_hidden=128, lstm_layers=2, dropout=0.3):
        super().__init__()
        self.stgcn = STGCN()
        self.lstm = nn.LSTM(128, lstm_hidden, lstm_layers, batch_first=True, bidirectional=True, dropout=dropout if lstm_layers > 1 else 0.0)
        self.proj = nn.Sequential(nn.Linear(lstm_hidden * 2, 128), nn.LayerNorm(128), nn.ReLU(), nn.Dropout(dropout))
        self.attention = SelfAttention(hidden_dim=128)
        self.classifier = nn.Sequential(nn.Linear(128 + 10, 64), nn.ReLU(), nn.Dropout(0.2), nn.Linear(64, n_classes))

    def encode(self, keypoints, adj, return_attention=False):
        if adj.dim() == 3: adj = adj[0]
        frame_feats = self.stgcn(keypoints, adj)
        lstm_out, _ = self.lstm(frame_feats)
        projected   = self.proj(lstm_out)
        context, attn_weights = self.attention(projected)
        return (context, attn_weights) if return_attention else (context, None)

    def forward(self, keypoints, adj, scalars):
        context, _ = self.encode(keypoints, adj)
        combined   = torch.cat([context, scalars], dim=1)
        return self.classifier(combined)

# ─── LSTM-only baseline ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

class LSTMOnly(nn.Module):
    def __init__(self, n_classes=2, lstm_hidden=128, lstm_layers=2, dropout=0.3):
        super().__init__()
        self.stgcn = STGCN()
        self.lstm = nn.LSTM(128, lstm_hidden, lstm_layers, batch_first=True, bidirectional=True, dropout=dropout if lstm_layers > 1 else 0.0)
        self.proj = nn.Sequential(nn.Linear(lstm_hidden * 2, 128), nn.LayerNorm(128), nn.ReLU(), nn.Dropout(dropout))
        self.classifier = nn.Sequential(nn.Linear(128 + 10, 64), nn.ReLU(), nn.Dropout(0.2), nn.Linear(64, n_classes))

    def encode(self, keypoints, adj, return_attention=False):
        if adj.dim() == 3: adj = adj[0]
        frame_feats = self.stgcn(keypoints, adj)
        lstm_out, _ = self.lstm(frame_feats)
        last_hidden  = self.proj(lstm_out[:, -1, :])
        return last_hidden, None

    def forward(self, keypoints, adj, scalars):
        context, _ = self.encode(keypoints, adj)
        combined   = torch.cat([context, scalars], dim=1)
        return self.classifier(combined)

def train_one_epoch_generic(model, loader, optimizer, device, class_weights=None) -> float:
    model.train()
    total_loss, used_batches = 0.0, 0
    A = torch.from_numpy(ADJ).to(device)
    for batch in loader:
        kps, labels, sc = batch['keypoints'].to(device), batch['label'].to(device), batch['scalars'].to(device)
        good = torch.isfinite(kps).flatten(1).all(1) & torch.isfinite(sc).all(1)
        if not good.any(): continue
        logits, _ = model(kps[good], A, sc[good]) if isinstance(model, RehabNet) else (model(kps[good], A, sc[good]), None)
        loss = F.cross_entropy(logits, labels[good], weight=class_weights)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        total_loss += loss.item(); used_batches += 1
    return total_loss / max(used_batches, 1)

@torch.no_grad()
def evaluate_generic(model, loader, device) -> dict:
    model.eval()
    A = torch.from_numpy(ADJ).to(device)
    all_preds, all_labels = [], []
    for batch in loader:
        kps, labels, sc = batch['keypoints'].to(device), batch['label'].to(device), batch['scalars'].to(device)
        good = torch.isfinite(kps).flatten(1).all(1) & torch.isfinite(sc).all(1)
        if not good.any(): continue
        logits, _ = model(kps[good], A, sc[good]) if isinstance(model, RehabNet) else (model(kps[good], A, sc[good]), None)
        all_preds.extend(logits.argmax(dim=1).cpu().numpy())
        all_labels.extend(labels[good].cpu().numpy())
    if not all_labels: return {'accuracy': 0, 'f1_macro': 0, 'incorrect_recall': 0}
    acc = accuracy_score(all_labels, all_preds)
    f1  = f1_score(all_labels, all_preds, average='macro', zero_division=0)
    rec = f1_score(all_labels, all_preds, average=None, zero_division=0)[0]
    return {'accuracy': acc, 'f1_macro': f1, 'incorrect_recall': rec}

def run_loso_single(model_class, model_name, samples, n_epochs=30, batch_size=32, lr=1e-3, device_str='cuda' if torch.cuda.is_available() else 'cpu'):
    device = torch.device(device_str)
    subjects = get_uiprmd_subjects(samples)
    results = []
    print(f"\nTraining: {model_name}")
    for subj in subjects:
        train_s, test_s = loso_split(samples, subj)
        train_dl, test_dl = build_loaders(train_s, test_s, batch_size)
        cw = compute_class_weights(train_s, device)
        model = model_class().to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=lr)
        for epoch in range(n_epochs):
            train_one_epoch_generic(model, train_dl, optimizer, device, cw)
        metrics = evaluate_generic(model, test_dl, device)
        results.append(metrics)
        print(f"  {subj} acc={metrics['accuracy']:.3f} f1={metrics['f1_macro']:.3f}")
    return results

def run_ablation(n_epochs=30, batch_size=32, lr=1e-3):
    samples = build_master_dataset()
    all_results = {}
    models = [(RehabNet, 'Transformer'), (SARNN, 'SA-RNN'), (LSTMOnly, 'LSTM-only')]
    for m_class, name in models:
        all_results[name] = run_loso_single(m_class, name, samples, n_epochs, batch_size, lr)
    print("\nABLATION TABLE")
    for name, res in all_results.items():
        acc = np.mean([r['accuracy'] for r in res])
        f1 = np.mean([r['f1_macro'] for r in res])
        print(f"{name:<15}: Acc={acc:.3f}, F1={f1:.3f}")
    return all_results
```

```python
"""
=============================================================================
PS2 — XGBoost Layer + Transformer Integration
=============================================================================
Run order in Colab:
  1. Paste + run ps2_pipeline.py  (defines RehabNet, build_master_dataset etc.)
  2. Paste + run this file        (defines run_full_pipeline, predict_one_rep)
  3. Run:
       trained_models, data = run_full_pipeline(
           csv_path='/content/drive/MyDrive/Datasets/final_cleaned_data.csv',
           transformer_model=None,   # baseline — swap in rehab_net later
       )
=============================================================================
"""

import numpy as np
import pandas as pd
import pickle, os, gzip
import warnings
warnings.filterwarnings('ignore')

try:
    import xgboost as xgb
    import shap
except ImportError:
    import subprocess, sys
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'xgboost', 'shap', '-q'])
    import xgboost as xgb
    import shap

from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.metrics import f1_score


# =============================================================================
# SECTION 1 — CONFIGURATION
# =============================================================================

CSV_PATH  = '/content/drive/MyDrive/datasets/final_cleaned_data.csv'
SAVE_DIR  = '/content/drive/MyDrive/datasets/'

# Only trunk_compensation has real signal (8.8% diff correct vs incorrect)
# insufficient_ROM and too_slow both collapsed to ~3% diff across full dataset
ERROR_CLASSES = [
    'err_trunk_compensation',
]

# Always drop these — identifiers, leaky, or all-null
DROP_ALWAYS = [
    'rep_id', 'exercise', 'n_frames_raw', 'subject_id', 'target_label',
    'source',
    'valgus_angle', 'err_trunk_compensation',        # r=-0.80 with label — binary flag not measurement
    'imu_accel_rms', 'imu_accel_peak', 'imu_accel_jerk_rms',
    'imu_gyro_rms', 'imu_gyro_peak',
    'imu_sway_ap', 'imu_sway_ml', 'imu_step_symm',
    'left_knee_stability_slr', 'left_knee_min_during_slr',
    'right_knee_stability_slr','right_knee_min_during_slr',
]

# Drop any column whose name contains these strings
DROP_PATTERNS = [
    'start_angle', 'end_angle', 'angle_mean', 'angle_std',  # leaky angle stats
    'hesitations',                                            # all identical value
    'flag_asymmetric',                                        # binary flag
    'left_knee_insufficient', 'right_knee_insufficient',
    'left_hip_insufficient',  'right_hip_insufficient',
    'left_knee_peak_not',     'right_knee_peak_not',
    'left_hip_peak_not',      'right_hip_peak_not',     'left_trunk_lean_proxy',    # ← ADD
    'right_trunk_lean_proxy',   # ← ADD
    'left_knee_too_',         'right_knee_too_',
    'left_hip_too_',          'right_hip_too_',
    'left_knee_high_',        'right_knee_high_',
    'left_hip_high_',         'right_hip_high_',
    'left_trunk_compensation','right_trunk_compensation',
    'knee_asymmetric',        'hip_asymmetric', 'ankle_asymmetric',
    'err_too_fast',           'err_high_jerk',  'err_asymmetric',  # inverted/broken
    'stability_slr',          'min_during_slr',
]

# Feedback message templates keyed by error class and pain level
FEEDBACK_TEMPLATES = {
    'err_trunk_compensation': {
        'low_pain':  "Keep your back straight — try not to lean forward during the movement.",
        'high_pain': "Focus on keeping your back upright — lean forward as little as possible.",
        'priority':  1,
    },
    'correct': {
        'low_pain':  "Great rep — keep it up!",
        'high_pain': "Well done — good movement within your comfortable range.",
        'priority':  0,
    },
}


# =============================================================================
# SECTION 2 — DATA LOADING
# =============================================================================

def load_and_clean(csv_path: str) -> dict:
    """
    Load CSV, drop leaky/broken/null columns, return clean feature matrix.
    """
    df = pd.read_csv(csv_path, low_memory=False)
    print(f"Loaded: {df.shape}")

    # Build drop set
    drop = set(DROP_ALWAYS)
    for col in df.columns:
        for pat in DROP_PATTERNS:
            if pat in col:
                drop.add(col)
        # Drop error columns we are NOT predicting
        if col.startswith('err_') and col not in ERROR_CLASSES:
            drop.add(col)

    # Only keep numeric columns not in drop set
    numeric_cols = set(df.select_dtypes(include=[np.number]).columns)
    feat_cols = [c for c in df.columns if c not in drop and c in numeric_cols]

    print(f"Clean features: {len(feat_cols)}")

    X_scalar  = df[feat_cols].fillna(0).values.astype(np.float32)
    y_binary  = df['target_label'].values.astype(int)
    y_errors  = df[ERROR_CLASSES].fillna(0).values.astype(int)  # (N, n_error_classes)
    subjects  = df['subject_id'].values

    scaler   = RobustScaler()
    X_scaled = scaler.fit_transform(X_scalar).astype(np.float32)

    le     = LabelEncoder()
    groups = le.fit_transform(subjects)

    return {
        'X_scalar':  X_scalar,
        'X_scaled':  X_scaled,
        'y_binary':  y_binary,
        'y_errors':  y_errors,
        'subjects':  subjects,
        'groups':    groups,
        'feat_cols': feat_cols,
        'scaler':    scaler,
        'df':        df,
    }


# =============================================================================
# SECTION 3 — BUILD COMBINED INPUT
# =============================================================================

def build_xgb_input(
    cls_embeddings: np.ndarray,  # (N, 128) — zeros if transformer not ready
    X_scaled:       np.ndarray,  # (N, n_feats)
) -> np.ndarray:
    """Concatenate CLS embeddings with scalar features → (N, 128+n_feats)."""
    return np.hstack([cls_embeddings, X_scaled]).astype(np.float32)


# =============================================================================
# SECTION 4 — XGBOOST TRAINING
# =============================================================================

def train_xgboost(
    X:         np.ndarray,   # (N, 128+n_feats) combined input
    y_binary:  np.ndarray,   # (N,) quality label
    y_errors:  np.ndarray,   # (N, n_error_classes)
    groups:    np.ndarray,   # (N,) subject groups for LOSO
    feat_cols: list,
) -> dict:
    """
    Train quality model (binary) + one XGBoost per error class.
    Returns trained models, SHAP explainers, LOSO F1 scores.
    """
    logo = LeaveOneGroupOut()
    n_features = X.shape[1]   # 128 + n_scalar — used for SHAP consistency

    # ── Quality model ─────────────────────────────────────────────────────────
    print("\n=== TRAINING QUALITY MODEL (binary correct/incorrect) ===")
    quality_params = dict(
        n_estimators=300, max_depth=6, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        eval_metric='logloss', random_state=42, n_jobs=-1,
    )

    loso_quality_f1 = []
    for tr, te in logo.split(X, y_binary, groups):
        m = xgb.XGBClassifier(**quality_params)
        m.fit(X[tr], y_binary[tr], eval_set=[(X[te], y_binary[te])], verbose=False)
        pred = m.predict(X[te])
        loso_quality_f1.append(f1_score(y_binary[te], pred, average='macro', zero_division=0))

    print(f"Quality LOSO F1 (macro): {np.mean(loso_quality_f1):.3f} ± {np.std(loso_quality_f1):.3f}")

    # Train final quality model on all data
    quality_model = xgb.XGBClassifier(**quality_params)
    quality_model.fit(X, y_binary, verbose=False)

    # ── Error class models ────────────────────────────────────────────────────
    print("\n=== TRAINING ERROR CLASS MODELS ===")
    error_models   = {}
    loso_error_f1  = {}

    for i, err_name in enumerate(ERROR_CLASSES):
        y_err = y_errors[:, i]
        n_pos = y_err.sum()
        n_neg = len(y_err) - n_pos
        spw   = n_neg / max(n_pos, 1)

        err_params = dict(
            n_estimators=200, max_depth=5, learning_rate=0.05,
            subsample=0.8, colsample_bytree=0.8,
            scale_pos_weight=spw,
            eval_metric='logloss', random_state=42, n_jobs=-1,
        )

        fold_f1 = []
        for tr, te in logo.split(X, y_err, groups):
            if y_err[te].sum() == 0:
                continue
            m = xgb.XGBClassifier(**err_params)
            m.fit(X[tr], y_err[tr], verbose=False)
            pred = m.predict(X[te])
            fold_f1.append(f1_score(y_err[te], pred, average='macro', zero_division=0))

        mean_f1 = np.mean(fold_f1) if fold_f1 else 0.0
        loso_error_f1[err_name] = mean_f1
        print(f"  {err_name:30s}: LOSO F1={mean_f1:.3f}  (positive rate={y_err.mean():.1%})")

        # Final model on all data
        err_model = xgb.XGBClassifier(**err_params)
        err_model.fit(X, y_err, verbose=False)
        error_models[err_name] = err_model

    # ── SHAP explainers — trained on same shape as final models ───────────────
    print("\nBuilding SHAP explainers...")
    quality_explainer = shap.TreeExplainer(quality_model)
    error_explainers  = {
        name: shap.TreeExplainer(model)
        for name, model in error_models.items()
    }

    # Build full feature name list for SHAP labelling
    all_feat_names = [f'cls_{i}' for i in range(128)] + feat_cols

    return {
        'quality_model':      quality_model,
        'error_models':       error_models,
        'quality_explainer':  quality_explainer,
        'error_explainers':   error_explainers,
        'loso_quality_f1':    loso_quality_f1,
        'loso_error_f1':      loso_error_f1,
        'feat_cols':          feat_cols,
        'all_feat_names':     all_feat_names,   # 128 CLS + scalar names
        'n_features':         n_features,
    }


# =============================================================================
# SECTION 5 — INFERENCE: SINGLE REP
# =============================================================================

def predict_one_rep(
    cls_embedding:   np.ndarray,  # (128,) CLS token — zeros if no transformer
    scalar_features: np.ndarray,  # (n_feats,) scaled scalar features
    trained_models:  dict,        # output of train_xgboost()
    pain_score:      int = 0,     # 0-10 NRS from patient
) -> dict:
    """
    Full inference for one repetition.
    Returns quality prediction, error flags, SHAP explanation, feedback message.
    """
    pain_level = 'high_pain' if pain_score >= 7 else 'low_pain'

    # Build combined input — must match training shape
    X_rep = np.hstack([cls_embedding, scalar_features]).reshape(1, -1).astype(np.float32)

    expected_n = trained_models['n_features']
    if X_rep.shape[1] != expected_n:
        raise ValueError(
            f"Feature mismatch: got {X_rep.shape[1]}, expected {expected_n}. "
            f"Make sure scalar_features has {expected_n - 128} elements."
        )

    # ── Quality prediction ────────────────────────────────────────────────────
    qm       = trained_models['quality_model']
    q_prob   = qm.predict_proba(X_rep)[0]
    q_pred   = int(qm.predict(X_rep)[0])
    q_conf_v = float(q_prob[q_pred])
    q_conf   = 'high' if q_conf_v > 0.80 else ('moderate' if q_conf_v > 0.60 else 'low')

    # ── Error predictions ─────────────────────────────────────────────────────
    errors_detected = {}
    for err_name, err_model in trained_models['error_models'].items():
        prob = float(err_model.predict_proba(X_rep)[0][1])
        pred = int(err_model.predict(X_rep)[0])
        errors_detected[err_name] = {
            'predicted':   pred,
            'probability': round(prob, 3),
            'confidence':  'high' if prob > 0.80 else ('moderate' if prob > 0.60 else 'low'),
        }

    # ── SHAP explanations — use full combined input ───────────────────────────
    all_feat_names   = trained_models['all_feat_names']
    shap_explanations = {}

    # Quality SHAP
    q_shap = trained_models['quality_explainer'].shap_values(X_rep)
    if isinstance(q_shap, list):
        q_shap = q_shap[1]   # positive class
    top3_idx = np.argsort(np.abs(q_shap[0]))[-3:][::-1]
    shap_explanations['quality'] = [
        {
            'feature':    all_feat_names[i] if i < len(all_feat_names) else f'feat_{i}',
            'shap_value': round(float(q_shap[0][i]), 4),
            'direction':  '↑ correct' if q_shap[0][i] > 0 else '↓ incorrect',
        }
        for i in top3_idx
    ]

    # Error SHAP (only for detected errors)
    for err_name in ERROR_CLASSES:
        if errors_detected[err_name]['predicted'] == 1:
            e_shap = trained_models['error_explainers'][err_name].shap_values(X_rep)
            if isinstance(e_shap, list):
                e_shap = e_shap[1]
            top3 = np.argsort(np.abs(e_shap[0]))[-3:][::-1]
            shap_explanations[err_name] = [
                {
                    'feature':    all_feat_names[i] if i < len(all_feat_names) else f'feat_{i}',
                    'shap_value': round(float(e_shap[0][i]), 4),
                }
                for i in top3
            ]

    # ── Patient feedback ──────────────────────────────────────────────────────
    active_errors = [
        err for err, info in errors_detected.items()
        if info['predicted'] == 1 and info['confidence'] != 'low'
    ]
    if active_errors:
        active_errors.sort(key=lambda e: FEEDBACK_TEMPLATES.get(e, {}).get('priority', 99))
        top_error    = active_errors[0]
        feedback_msg = FEEDBACK_TEMPLATES[top_error][pain_level]
    else:
        top_error    = None
        feedback_msg = FEEDBACK_TEMPLATES['correct'][pain_level]

    # ── Therapist report ──────────────────────────────────────────────────────
    report_lines = [
        f"Quality: {'CORRECT' if q_pred==1 else 'INCORRECT'} "
        f"(confidence={q_conf}, p={q_conf_v:.2f})"
    ]
    if active_errors:
        report_lines.append(f"Errors detected ({len(active_errors)}):")
        for name in active_errors:
            info = errors_detected[name]
            report_lines.append(f"  {name}: p={info['probability']:.2f} ({info['confidence']})")
            if name in shap_explanations:
                top = shap_explanations[name][0]
                feat_clean = top['feature'].replace('left_','').replace('_',' ')
                direction  = "high" if top['shap_value'] > 0 else "low"
                report_lines.append(f"    → primary driver: {feat_clean} ({direction})")
    else:
        report_lines.append("No errors detected.")

    return {
        'feedback_message':   feedback_msg,
        'feedback_error':     top_error,
        'quality_correct':    q_pred,
        'quality_prob':       round(q_prob[1], 3),
        'quality_confidence': q_conf,
        'errors_detected':    errors_detected,
        'shap_explanations':  shap_explanations,
        'therapist_report':   "\n".join(report_lines),
        'flag_for_review':    q_conf == 'low',
    }


# =============================================================================
# SECTION 6 — EXTRACT CLS EMBEDDINGS FROM TRAINED TRANSFORMER
# =============================================================================

def extract_cls_embeddings(
    model,           # trained RehabNet instance
    samples: list,   # output of build_master_dataset()
    batch_size: int = 64,
    device_str: str = 'cuda',
) -> np.ndarray:
    """
    Extract (N, 128) CLS embeddings from trained RehabNet.
    Call this after transformer LOSO training is done.

    Usage:
        samples = build_master_dataset(lower_limb_only=True, use_uco=True)
        cls_embs = extract_cls_embeddings(rehab_net, samples)
        # then pass cls_embs to run_full_pipeline() via transformer_model
    """
    import torch
    from ps2_pipeline import ADJ

    device = torch.device(device_str if torch.cuda.is_available() else 'cpu')
    model  = model.to(device)
    model.eval()

    adj_t = torch.from_numpy(ADJ).to(device)
    all_cls = []

    with torch.no_grad():
        for i in range(0, len(samples), batch_size):
            batch   = samples[i:i+batch_size]
            kps     = torch.stack([
                torch.from_numpy(s['keypoints']).permute(2, 0, 1)  # (3,T,6)
                for s in batch
            ]).to(device)   # (B, 3, T, 6)

            ctx, _ = model.encode(kps, adj_t, return_attention=False)
            all_cls.append(ctx.cpu().numpy())

    embeddings = np.vstack(all_cls)
    print(f"CLS embeddings extracted: {embeddings.shape}")
    return embeddings


# =============================================================================
# SECTION 7 — FULL PIPELINE
# =============================================================================

def run_full_pipeline(
    csv_path:          str  = CSV_PATH,
    transformer_model        = None,   # pass trained RehabNet to use real CLS
    samples:           list = None,    # pass build_master_dataset() output
    save_models:       bool = False,
) -> tuple:
    """
    Run complete PS2 pipeline.

    Without transformer (baseline):
        trained, data = run_full_pipeline(csv_path='...')

    With transformer (full system):
        samples = build_master_dataset(lower_limb_only=True, use_uco=True)
        trained, data = run_full_pipeline(
            csv_path='...',
            transformer_model=rehab_net,
            samples=samples,
        )
    """
    print("=" * 60)
    print("PS2 FULL PIPELINE")
    print("=" * 60)

    # Step 1 — load scalar features
    print("\n[1/5] Loading and cleaning data...")
    data = load_and_clean(csv_path)
    N    = len(data['X_scaled'])

    # Step 2 — get CLS embeddings
    print("\n[2/5] Getting CLS embeddings...")
    if transformer_model is not None and samples is not None:
        print("  Extracting real CLS embeddings from trained transformer...")
        cls_embeddings = extract_cls_embeddings(transformer_model, samples)
        if len(cls_embeddings) != N:
            print(f"  WARNING: transformer has {len(cls_embeddings)} samples, "
                  f"CSV has {N}. Using zeros.")
            cls_embeddings = np.zeros((N, 128), dtype=np.float32)
    else:
        print("  No transformer provided — using zero embeddings (baseline mode).")
        print("  Expected F1 ~0.635. Will improve when transformer is connected.")
        cls_embeddings = np.zeros((N, 128), dtype=np.float32)

    # Step 3 — build combined input
    print("\n[3/5] Building combined input...")
    X_combined = build_xgb_input(cls_embeddings, data['X_scaled'])
    print(f"  Shape: {X_combined.shape}  (128 CLS + {data['X_scaled'].shape[1]} scalar)")

    # Step 4 — train XGBoost
    print("\n[4/5] Training XGBoost models (LOSO)...")
    trained = train_xgboost(
        X_combined,
        data['y_binary'],
        data['y_errors'],
        data['groups'],
        data['feat_cols'],
    )

    # Step 5 — test inference on one rep
    print("\n[5/5] Testing inference on one rep...")
    idx = 42
    result = predict_one_rep(
        cls_embedding   = cls_embeddings[idx],
        scalar_features = data['X_scaled'][idx],
        trained_models  = trained,
        pain_score      = 2,
    )
    print(f"\n  Rep {idx} (true label={data['y_binary'][idx]}):")
    print(f"  → Patient: '{result['feedback_message']}'")
    print(f"  → Quality: {'CORRECT' if result['quality_correct'] else 'INCORRECT'} "
          f"({result['quality_confidence']} confidence)")
    print(f"  → Flag for review: {result['flag_for_review']}")
    print(f"\n  Therapist report:")
    for line in result['therapist_report'].split('\n'):
        print(f"    {line}")
    print(f"\n  Top SHAP drivers (quality):")
    for s in result['shap_explanations'].get('quality', []):
        print(f"    {s['feature']:35s} {s['shap_value']:+.4f}  {s['direction']}")

    # Save
    if save_models:
        save_path = os.path.join(SAVE_DIR, 'ps2_xgboost_models.pkl')
        with open(save_path, 'wb') as f:
            pickle.dump({
                'quality_model':     trained['quality_model'],
                'error_models':      trained['error_models'],
                'quality_explainer': trained['quality_explainer'],
                'error_explainers':  trained['error_explainers'],
                'scaler':            data['scaler'],
                'feat_cols':         data['feat_cols'],
                'all_feat_names':    trained['all_feat_names'],
                'loso_quality_f1':   trained['loso_quality_f1'],
                'loso_error_f1':     trained['loso_error_f1'],
            }, f)
        print(f"\n  Saved to {save_path}")

    # Summary
    print("\n" + "=" * 60)
    print("RESULTS SUMMARY")
    print("=" * 60)
    print(f"Quality LOSO F1 (macro): "
          f"{np.mean(trained['loso_quality_f1']):.3f} ± "
          f"{np.std(trained['loso_quality_f1']):.3f}")
    print("Error class LOSO F1:")
    for name, f1 in trained['loso_error_f1'].items():
        print(f"  {name:35s}: {f1:.3f}")
    if transformer_model is None:
        print("\n  NOTE: Baseline results (zero CLS embeddings).")
        print("  Connect transformer for +0.05-0.15 F1 improvement.")

    return trained, data


# =============================================================================
# ENTRY POINT
# =============================================================================

if __name__ == '__main__':
    CSV = CSV_PATH
    if not os.path.exists(CSV):
        CSV = '/content/drive/MyDrive/datasets/final_cleaned_data.csv'

    trained_models, data = run_full_pipeline(
        csv_path    = CSV,
        save_models = False,
    )
```
